# ការព្យាករណ៍ស៊េរីពេលវេលាមួយជាមួយ ARIMA

នៅក្នុងសៀវភៅកំណត់ត្រានេះ យើងបង្ហាញពីរបៀប៖
- រៀបចំទិន្នន័យស៊េរីពេលវេលាសម្រាប់បណ្តុះម៉ូឌែលព្យាករណ៍ស៊េរីពេលវេលា ARIMA
- អនុវត្តម៉ូឌែល ARIMA ងាយៗមួយ ដើម្បីព្យាករណ៍ជំហាន HORIZON បន្ទាប់មុខ (ពេល *t+1* រហូតដល់ *t+HORIZON*) ក្នុងស៊េរីពេលវេលា
- វាយតំលៃម៉ូឌែល
