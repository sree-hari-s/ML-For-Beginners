## бһҹбҹ’бһңбҹӮбһ„бһҷбһӣбҹӢбһўбҹҶбһ–бһёбһҖбһ¶бһҡбһ”бҹӮбһ„бһ…бҹӮбһҖбһҖбҹ’бһҡбһ»бһҳ K-Means бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ R бһ“бһ·бһ„бһӮбҹ„бһӣбһҖбһ¶бһҡбһҺбҹҚбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ Tidyбҹ” 

### [**бһҹбҹҶбһҺбһҪбһҡбһ–бһ·бһ…бһ¶бһҡбһҺбһ¶ бһҳбһ»бһ“бһҳбһ»бһҒбһңбһ·бһҮбҹ’бһҮбһ¶**](https://gray-sand-07a10f403.1.azurestaticapps.net/quiz/29/)

бһҖбҹ’бһ“бһ»бһ„бһҳбһ»бһҒбһңбһ·бһҮбҹ’бһҮбһ¶бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“бһ–бһёбһҡбһ”бҹҖбһ”бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбҹ’бһҡбһ»бһҳ cluster бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҖбһүбҹ’бһ…бһ”бҹӢ Tidymodels бһ“бһ·бһ„бһҖбһүбҹ’бһ…бһ”бҹӢбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһҖбҹ’бһ“бһ»бһ„бһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһўбһ·бһҒбһјбһҹбҹҠбһёбһҹбҹ’бһ‘бһ№бһҳ R (бһҷбһҫбһ„бһ“бһ№бһ„бһ бҹ…бһ–бһҪбһҖбһңбһ¶бһҗбһ¶ бһҳбһ·бһҸбҹ’бһҸ рҹ§‘вҖҚрҹӨқвҖҚрҹ§‘) бһ“бһ·бһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҸбһ“бҹ’бһҡбҹ’бһҸбһёбһ“бҹғбһ”бҹ’бһҡбһ‘бҹҒбһҹбһ“бһёбһ бҹ’бһҹбҹҒбһҡбһёбһҷбҹүбһ¶бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһ“бһ¶бҹҶбһ…бһјбһӣбһҳбһ»бһ“бһ“бҹҒбҹҮбҹ” бһҷбһҫбһ„бһ“бһ№бһ„бһӮбҹ’бһҡбһ”бһҠбһҺбҹ’бһҸбһ”бҹӢбһ–бһёбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһҡбһ”бһҹбҹӢ K-Means бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ”бҹӮбһ„бһ…бҹӮбһҖбһҖбҹ’бһҡбһ»бһҳбҹ” бһҹбһјбһҳбһ…бҹҶбһҺбһ¶бҹҶбһҗбһ¶ бһ…бҹҶбһ“бһ»бһ…бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһҡбҹҖбһ“бһҖбҹ’бһ“бһ»бһ„бһҳбһ»бһҒбһңбһ·бһҮбҹ’бһҮбһ¶бһҳбһ»бһ“ бһҳбһ¶бһ“бһҡбһ”бҹҖбһ”бһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷбһҖбҹ’бһҡбһ»бһҳ бһ бһҫбһҷбһңбһ·бһ’бһёбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһўбһ¶бһҹбҹ’бһҡбҹҗбһҷбһӣбһҫбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбҹ” бһҷбһҫбһ„бһ“бһ№бһ„бһҹбһ¶бһҖбһӣбҹ’бһ”бһ„ K-Means бһ–бҹ’бһҡбҹ„бҹҮбһңбһ¶бһҮбһ¶бһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡбһ”бҹӮбһ„бһ…бҹӮбһҖбһҖбҹ’бһҡбһ»бһҳбһҠбҹӮбһӣбһ–бҹҒбһүбһ“бһ·бһҷбһҳбһ”бҹҶбһ•бһ»бһҸбҹ” бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһ‘бҹ…!

бһ–бһ¶бһҖбҹ’бһҷбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“бҹ–

-   бһҖбһ¶бһҡбһңбһ¶бһҷбһҸбһҳбҹ’бһӣбҹғ Silhouette

-   бһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҡбҹ’бһҸ Elbow

-   бһҖбһҳбҹ’бһӣбһ¶бҹҶбһ„бһҡбһ№бһ„ (Inertia)

-   бһ—бһ¶бһ–бһ…бһҳбҹ’бһҡбһ»бҹҮ (Variance)

### **бһҖбһ¶бһҡбһҺбҹӮбһ“бһ¶бҹҶ**

[K-Means Clustering](https://wikipedia.org/wiki/K-means_clustering) бһӮбһәбһҮбһ¶бһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡбһҠбҹӮбһӣбһ”бһ¶бһ“бһ•бҹ’бһҹбҹҶбһ“бһ№бһ„бһ–бһёбһҠбҹӮбһ“бһҹбһүбҹ’бһүбһ¶бһҹбҹ’бһҡбһ¶бһңбһҮбҹ’бһҡбһ¶бһңбҹ” бһңбһ¶бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бҹӮбһ„бһ…бҹӮбһҖбһ“бһ·бһ„бһ…бҹӮбһҖбһҮбһ¶бһҖбҹ’бһҡбһ»бһҳбһ“бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ‘бҹ…бһҮбһ¶ `k clusters` бһҠбҹ„бһҷбһ•бҹ’бһўбҹӮбһҖбһӣбһҫбһҹбһҳбһ¶бһ’бһёбһ“бҹ…бһӣбһҫбһӣбһҖбҹ’бһҒбһҺбҹҲбһ–бһ·бһҹбҹҒбһҹбһҡбһ”бһҹбҹӢбһңбһ¶бҹ” 

бһҖбҹ’бһҡбһ»бһҳ cluster бһўбһ¶бһ…бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бһ„бҹ’бһ бһ¶бһүбһҮбһ¶бһҡбһјбһ”бһ—бһ¶бһ– [Voronoi diagrams](https://wikipedia.org/wiki/Voronoi_diagram) бһҠбҹӮбһӣбһҡбһҪбһҳбһҳбһ¶бһ“бһ…бҹҶбһҺбһ»бһ…бһҳбһҪбһҷ (бһ¬ 'бһӮбҹ’бһҡбһ¶бһ”бҹӢбһ–бһјбһҮбһ¶') бһ“бһ·бһ„бһҸбҹҶбһ”бһ“бҹӢбһҠбҹӮбһӣбһ–бһ¶бһҖбҹӢбһ–бҹҗбһ“бҹ’бһ’бһҡбһ”бһҹбҹӢбһңбһ¶бҹ”

бһҖбһ¶бһҡбһ”бҹӮбһ„бһ…бҹӮбһҖбһҖбҹ’бһҡбһ»бһҳ K-Means бһҳбһ¶бһ“бһҮбҹҶбһ бһ¶бһ“бһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

1.  бһўбҹ’бһ“бһҖбһңбһ·бһ‘бҹ’бһҷбһ¶бһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһҠбҹ„бһҷбһҖбҹҶбһҺбһҸбҹӢбһ…бҹҶбһ“бһҪбһ“бһҖбҹ’бһҡбһ»бһҳбһҠбҹӮбһӣбһ…бһ„бҹӢбһ”бһ„бҹ’бһҖбһҫбһҸбҹ”

2.  бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖ, бһўбһ¶бһӣбҹӢбһ бҹ’бһӮбһјбһҡбһ·бһ’бһҳбҹҚбһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҡбһ¶бһ”бҹӢ K бһўбһ¶бһҡбһҳбҹ’бһҳбһҺбҹҚбһҮбһ¶бһ”бҹӢбһ…бҹғбһҠбһ“бҹ’бһҷбһ–бһёбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһҳбҹ’бһҡбһҫбһҮбһ¶бһҳбһҮбҹ’бһҲбһҳбһҺбҹ’бһҢбһӣбһҠбҹҶбһ”бһјбһ„ (centroids) бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбҹ’бһҡбһ»бһҳ cluster бҹ”

3.  бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖ, бһҡбһ¶бһӣбҹӢбһўбһ¶бһҡбһҳбҹ’бһҳбһҺбҹҚбһ“бҹ…бһҹбһӣбҹӢбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ…бһ¶бһҸбҹӢбһ‘бһ»бһҖбһ‘бҹ…бһҖбһ¶бһ“бҹӢ centroid бһҠбҹӮбһӣбһҮбһ·бһҸбһ”бҹҶбһ•бһ»бһҸбҹ”

4.  бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖ, бһӮбһҺбһ“бһ¶бһҳбһ’бҹ’бһҷбһҳбһҗбҹ’бһҳбһёбһ“бҹғбһҡбһ¶бһӣбҹӢбһҖбҹ’бһҡбһ»бһҳ бһ бһҫбһҷбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ‘бһё centroid бһ‘бҹ…бһҖбһ¶бһ“бҹӢбһҳбһ’бҹ’бһҷбһҳбһ“бҹ„бҹҮбҹ”

5.  бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҠбҹӮбһӣбһҳбһҮбҹ’бһҲбһҳбһҺбҹ’бһҢбһӣбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһӮбһҺбһ“бһ¶бһЎбһҫбһ„бһңбһ·бһү бһўбһ¶бһҡбһҳбҹ’бһҳбһҺбҹҚбһҡбһ¶бһӣбҹӢбһҳбһҪбһҷбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҳбҹ’бһҸбһ„бһ‘бҹҖбһҸбһҗбһ¶бһҸбһҫбһңбһ¶бһўбһ¶бһ…бһҮбһ·бһҸбһҖбҹ’бһҡбһ»бһҳбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһҒбҹ’бһӣбҹҮбһ¬бһ‘бҹҒбҹ” бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһ…бһ¶бһҸбҹӢбһ‘бһ»бһҖбһңбһ¶бһЎбһҫбһ„бһңбһ·бһүбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҳбһ’бҹ’бһҷбһҳбһҖбҹ’бһҡбһ»бһҳбһҗбҹ’бһҳбһёбҹ” бһҮбҹҶбһ бһ¶бһ“бһ…бһ¶бһҸбҹӢбһҸбһ¶бҹҶбһ„бһҖбҹ’бһҡбһ»бһҳ бһ“бһ·бһ„бһ’бҹ’бһңбһҫбһ”бһ…бҹ’бһ…бһ»бһ”бҹ’бһ”бһ“бҹ’бһ“бһ—бһ¶бһ– centroid бһ“бһ№бһ„бһ’бҹ’бһңбһҫбһЎбһҫбһ„бһҮбһ¶бһҡбһҷбҹҲбһ–бҹҒбһӣбһҳбһҪбһҷбһҡбһ бһјбһҸбһҠбһӣбҹӢбһҖбһ¶бһҡбһ”бҹӮбһ„бһ…бҹӮбһҖбһҖбҹ’бһҡбһ»бһҳбһҲбһ”бҹӢбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡ (бһ§. бһ“бҹ…бһ–бҹҒбһӣбһҠбҹӮбһӣбһҖбһ¶бһҡбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһ”бһ¶бһ“бһҮбҹ„бһӮбһҮбҹҗбһҷ)бҹ” бһҮбһ¶бһ‘бһјбһ‘бҹ… бһўбһ¶бһӣбҹӢбһ бҹ’бһӮбһјбһҡбһ·бһ’бһҳбҹҚбһ“бһ№бһ„бһ”бһүбҹ’бһ…бһ”бҹӢбһ“бҹ…бһ–бҹҒбһӣбһҡбҹҖбһ„бһҡбһ¶бһӣбҹӢбһңбһҠбҹ’бһҸбһҗбҹ’бһҳбһёбһҳбһ¶бһ“бһ…бһӣбһ“бһ¶бһҸбһ·бһ…бһЎбһҫбһ„бһ“бҹғбһ…бҹҶбһҺбһ»бһ…бһҖбһҺбҹ’бһҸбһ¶бһӣ бһ бһҫбһҷбһҖбҹ’бһҡбһ»бһҳбһҖбҹ’бһӣбһ¶бһҷбһ‘бҹ…бһҮбһ¶бһҹбҹ’бһҗбһ·бһҸбһ·бҹ”

бһ“бҹҒбҹҮбһҮбһ¶бһҖбһ¶бһҡбһҹбһ„бҹ’бһҒбҹҒбһ”бһўбҹҶбһ–бһёбһўбҹ’бһңбһёбһҠбҹӮбһӣбһҷбһҫбһ„бһ”бһ¶бһ“бһ’бҹ’бһңбһҫбһ“бҹ…бһҗбҹ’бһ“бһ¶бһҖбҹӢбһҳбһ»бһ“бҹ” бһҷбһҫбһ„бһ…бҹӮбһҖбһ“бһ·бһ„бһ…бҹҶбһ бһ¶бһ…бҹӮбһҖбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҳбһҪбһҷбһ…бҹҶбһ“бһҪбһ“!


```r
# Load the core tidyverse and make it available in your current R session
library(tidyverse)

# Import the data into a tibble
df <- read_csv(file = "https://raw.githubusercontent.com/microsoft/ML-For-Beginners/main/5-Clustering/data/nigerian-songs.csv", show_col_types = FALSE)

# Narrow down to top 3 popular genres
nigerian_songs <- df %>% 
  # Concentrate on top 3 genres
  filter(artist_top_genre %in% c("afro dancehall", "afropop","nigerian pop")) %>% 
  # Remove unclassified observations
  filter(popularity != 0)



# Visualize popular genres using bar plots
theme_set(theme_light())
nigerian_songs %>%
  count(artist_top_genre) %>%
  ggplot(mapping = aes(x = artist_top_genre, y = n,
                       fill = artist_top_genre)) +
  geom_col(alpha = 0.8) +
  paletteer::scale_fill_paletteer_d("ggsci::category10_d3") +
  ggtitle("Top genres") +
  theme(plot.title = element_text(hjust = 0.5))
```

бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ“бҹҒбҹҮбһҹбҹ’бһўбһ¶бһҸбһ”бҹүбһ»бһҺбҹ’бһҺбһ¶бһҠбҹӮбһҡ? бһҷбһ¶бһ„бһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢ outliers бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ box plotsбҹ” бһ–бһҪбһҖбһҷбһҫбһ„бһ“бһ№бһ„бһ•бҹ’бһҠбҹ„бһҸбһӣбһҫбһҮбһҪбһҡбһҲбһҡбһ•бҹ’бһ“бҹӮбһҖбһӣбҹҒбһҒбһҠбҹӮбһӣбһҳбһ¶бһ“ outliers бһҸбһ·бһ…бһҮбһ¶бһ„ (бһ”бһҫбһ‘бҹ„бҹҮбһ”бһёбһҮбһ¶бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҹбһҳбҹ’бһўбһ¶бһҸ outliers бһ”бһ¶бһ“бһҖбҹҸбһҠбҹ„бһҷ)бҹ” Boxplots бһўбһ¶бһ…бһ”бһ„бҹ’бһ бһ¶бһүбһҮбһҪбһҡбһ“бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ бһ“бһ·бһ„бһҮбһҪбһҷбһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҮбһҪбһҡбһҲбһҡбһҺбһ¶бһҠбһҫбһҳбҹ’бһ”бһёбһ”бҹ’бһҡбһҫбҹ” бһҹбһҳбҹ’бһӮбһ¶бһӣбҹӢ, Boxplots бһҳбһ·бһ“бһ”бһ„бҹ’бһ бһ¶бһүбһ—бһ¶бһ–бһ•бҹ’бһҹбҹҒбһ„бһӮбҹ’бһ“бһ¶бһ‘бҹҒ, бһҠбҹӮбһӣбһҮбһ¶бһ’бһ¶бһҸбһ»бһҹбҹҶбһҒбһ¶бһ“бҹӢбһ“бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһўбһ¶бһ…бһ”бҹӮбһ„бһ…бҹӮбһҖбһ”бһ¶бһ“бһӣбҹ’бһўбҹ” бһҹбһјбһҳбһҳбһҫбһӣ [бһҖбһ¶бһҡбһ–бһ·бһ—бһ¶бһҖбҹ’бһҹбһ¶бһ“бҹҒбҹҮ](https://stats.stackexchange.com/questions/91536/deduce-variance-from-boxplot) бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһўбһ¶бһ“бһ”бһ“бҹ’бһҗбҹӮбһҳбҹ”

[Boxplots](https://en.wikipedia.org/wiki/Box_plot) бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бҹ’бһҡбһҫбһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһ бһ¶бһүбһ—бһ¶бһ–бһ…бҹӮбһҖбһ…бһ¶бһҷбһҷбҹүбһ¶бһ„бһҖбҹ’бһҡбһ¶бһ бҹ’бһңбһ·бһҖбһ“бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ `numeric` бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮ бһҷбһҫбһ„бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһҠбҹ„бһҷ *бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹ* бһҮбһҪбһҡбһҲбһҡбһӣбҹҒбһҒбһ‘бһ¶бҹҶбһ„бһўбһҹбҹӢбһҡбһҪбһҳбһҮбһ¶бһҳбһҪбһҷбһ”бҹ’бһҡбһ—бҹҒбһ‘бһҸбһ“бҹ’бһҸбҹ’бһҡбһёбһ–бҹҒбһүбһ“бһ·бһҷбһҳбҹ”


```r
# Select top genre column and all other numeric columns
df_numeric <- nigerian_songs %>% 
  select(artist_top_genre, where(is.numeric)) 

# Display the data
df_numeric %>% 
  slice_head(n = 5)
```

бһҳбһҫбһӣбһҗбһ¶вҖӢбһҮбһҪбһҷвҖӢбһҖбһ¶бһҡбһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹ `where` бһ’бҹ’бһңбһҫбһІбҹ’бһҷвҖӢбһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣвҖӢбһ“бҹҒбҹҮ рҹ’Ғ? бһҹбҹ’бһңбҹӮбһ„бһҷбһӣбҹӢвҖӢбһўбҹҶбһ–бһёвҖӢбһҳбһ»бһҒбһ„бһ¶бһҡвҖӢбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸвҖӢбһ‘бһёбһ“бҹҒбҹҮ [here](https://tidyselect.r-lib.org/)бҹ”

бһҠбҹ„бһҷбһҹбһ¶бһҡвҖӢбһҷбһҫбһ„вҖӢбһ“бһ№бһ„вҖӢбһ”бһ„бҹ’бһҖбһҫбһҸвҖӢбһҖбҹҶбһҺбһҸбҹӢвҖӢбһ”бһ№бһ„вҖӢбһҡбһ¶бһ„вҖӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢвҖӢбһӣбһҖбҹ’бһҒбһҺбҹҲбһ…бҹҶбһ“бһҪбһ“вҖӢбһҡбһ¶бһӣбҹӢбһҳбһҪбһҷ бһ бһҫбһҷвҖӢбһҷбһҫбһ„вҖӢбһ…бһ„бҹӢбһ…бҹҖбһҹбһңбһ¶бһ„вҖӢбһҖбһ¶бһҡвҖӢбһ”бҹ’бһҡбһҫбһӣбҹҶбһ бһ¶бһҸбҹӢвҖӢбһҶбҹ’бһ“бһ¶бҹҶбһЎбһјбһ” бһҹбһјбһҳбһ”бҹ’бһҡбҹӮвҖӢбһ‘бҹ’бһҡбһ„бҹӢвҖӢбһ‘бҹ’бһҡбһ¶бһҷвҖӢбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷвҖӢбһҡбһ”бһҹбҹӢвҖӢбһҷбһҫбһ„вҖӢбһ‘бҹ…бһҮбһ¶вҖӢбһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷ *бһңбҹӮбһ„* бһҠбҹӮбһӣвҖӢбһ“бһ№бһ„бһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸвҖӢбһІбҹ’бһҷвҖӢбһҷбһҫбһ„вҖӢбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ `facets` - бһӮбҹҶбһ“бһјбһҹвҖӢбһҡбһјбһ”вҖӢбһҸбһјбһ…вҖӢбһҠбҹӮбһӣвҖӢбһҳбһҪбһҷбһҡбһјбһ”вҖӢбһ”бһ„бҹ’бһ бһ¶бһүвҖӢбһІбҹ’бһҷвҖӢбһғбһҫбһүвҖӢбһҠбһ¶бһҖбҹӢвҖӢбһ…бҹҒбһүвҖӢбһҸбҹҶбһ”бһ“бҹӢвҖӢбһҳбһҪбһҷвҖӢбһ“бҹғвҖӢбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбҹ”


```r
# Pivot data from wide to long
df_numeric_long <- df_numeric %>% 
  pivot_longer(!artist_top_genre, names_to = "feature_names", values_to = "values") 

# Print out data
df_numeric_long %>% 
  slice_head(n = 15)
```

бһңбҹӮбһ„бһҮбһ¶бһ„бһ“бҹҒбҹҮбһ‘бҹҖбһҸ! бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһ–бҹҒбһӣбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢ `ggplots`! бһҸбһҫбһҷбһҫбһ„бһ“бһ№бһ„бһ”бҹ’бһҡбһҫ `geom` бһўбҹ’бһңбһё?


```r
# Make a box plot
df_numeric_long %>% 
  ggplot(mapping = aes(x = feature_names, y = values, fill = feature_names)) +
  geom_boxplot() +
  facet_wrap(~ feature_names, ncol = 4, scales = "free") +
  theme(legend.position = "none")
```

Easy-gg!

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҷбһҫбһ„бһўбһ¶бһ…бһҳбһҫбһӣбһғбһҫбһүбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ“бҹҒбҹҮбһҳбһ¶бһ“бһҹбҹҶбһЎбҹҒбһ„бһҡбҹҶбһҒбһ¶бһ“бһ”бһ“бҹ’бһҸбһ·бһ…бҹ– бһҠбҹ„бһҷбһҖбһ¶бһҡбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҳбһҫбһӣбһҮбһҪбһҡбһҠбҹҒбһҖбһ“бһёбһҳбһҪбһҷбҹ—бһҮбһ¶бһҡбһ¶бһ„бһ”бҹ’бһҡбһўбһ”бҹӢ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһғбһҫбһүбһҸбһҳбҹ’бһӣбҹғбһ…бҹҶбһҳбҹ’бһӣбһ„бҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҸбһ¶бһҡбһ¶бһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ бһҫбһҷбһҠбһҖбһҸбһҳбҹ’бһӣбҹғбһ…бҹҶбһҳбҹ’бһӣбһ„бһ‘бһ¶бҹҶбһ„бһ“бҹҒбҹҮбһ…бҹҒбһү бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһңбһ¶бһ“бһ№бһ„бһ’бҹ’бһңбһҫбһұбҹ’бһҷбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҳбһ¶бһ“бһҸбһ·бһ…бһҸбһҪбһ…бҹ”

бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ–бҹҒбһӣбһ“бҹҒбҹҮ бһҷбһҫбһ„бһҳбһҖбһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҮбһҪбһҡбһҠбҹҒбһҖбһҠбҹӮбһӣбһҷбһҫбһ„бһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ бһ¶бһҸбҹӢ clustering бһҡбһ”бһҹбҹӢбһҷбһҫбһ„бҹ” бһҷбһҫбһ„бһ“бһ№бһ„бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҮбһҪбһҡбһҠбҹҒбһҖбһҮбһ¶бһ…бҹҶбһ“бһҪбһ“бһҠбҹӮбһӣбһҳбһ¶бһ“бһҮбһҪбһҡбһҠбһјбһ…бһӮбҹ’бһ“бһ¶бҹ” бһҷбһҫбһ„бһўбһ¶бһ…бһ”бһҳбҹ’бһӣбҹӮбһ„ `artist_top_genre` бһҮбһ¶бһ…бҹҶбһ“бһҪбһ“бһҸбҹӮбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ–бҹҒбһӣбһ“бҹҒбҹҮбһҷбһҫбһ„бһ“бһ№бһ„бһӣбһ»бһ”бһңбһ¶бһ…бҹҒбһүбҹ”


```r
# Select variables with similar ranges
df_numeric_select <- df_numeric %>% 
  select(popularity, danceability, acousticness, loudness, energy) 

# Normalize data
# df_numeric_select <- scale(df_numeric_select)
```

## 3. бһҖбҹҶбһҺбһҸбҹӢбһҖбҹ’бһҡбһ»бһҳбһҸбһ¶бһҳбһңбһ·бһ’бһё k-means бһҖбҹ’бһ“бһ»бһ„ R

бһҷбһҫбһ„бһўбһ¶бһ…бһҖбҹҶбһҺбһҸбҹӢбһҖбҹ’бһҡбһ»бһҳбһҸбһ¶бһҳбһңбһ·бһ’бһё k-means бһҖбҹ’бһ“бһ»бһ„ R бһҮбһ¶бһҳбһҪбһҷбһҳбһ»бһҒбһ„бһ¶бһҡ `kmeans` бһҠбҹӮбһӣбһҳбһ¶бһ“бһҳбһ»бһ“бһ бһҫбһҷ бһҹбһјбһҳбһҳбһҫбһӣ `help("kmeans()")`бҹ” бһҳбһ»бһҒбһ„бһ¶бһҡ `kmeans()` бһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“ data frame бһҠбҹӮбһӣбһҳбһ¶бһ“бһҮбһҪбһҡбһҲбһҡбһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҒбҹ’бһӣбһҪбһ“бһ‘бһ¶бҹҶбһ„бһўбһҹбҹӢбһҮбһ¶бһҲбҹ’бһҳбҹ„бҹҮбһ”бһүбҹ’бһ…бһјбһӣбһҠбһҫбһҳбһҡбһ”бһҹбҹӢбһңбһ¶бҹ”

бһҮбҹҶбһ бһ¶бһ“бһҠбҹҶбһ”бһјбһ„ бһ“бҹ…бһ–бҹҒбһӣбһ”бҹ’бһҡбһҫбһҖбҹ’бһҡбһ»бһҳбһҸбһ¶бһҳбһңбһ·бһ’бһё k-means бһӮбһәбһҸбҹ’бһҡбһјбһңбһҖбҹҶбһҺбһҸбҹӢбһ…бҹҶбһ“бһҪбһ“бһҖбҹ’бһҡбһ»бһҳ (k) бһҠбҹӮбһӣбһ“бһ№бһ„бһҸбҹ’бһҡбһјбһңбһ”бһ„бҹ’бһҖбһҫбһҸбһЎбһҫбһ„бһҖбҹ’бһ“бһ»бһ„бһҠбҹҶбһҺбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷбҹ” бһҷбһҫбһ„бһҠбһ№бһ„бһҗбһ¶бһҳбһ¶бһ“ 3 бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ”бһ‘бһ…бһҳбҹ’бһҡбҹҖбһ„бһҠбҹӮбһӣбһҷбһҫбһ„бһ”бһ¶бһ“бһ…бҹӮбһ„бһ…бҹҒбһүбһ–бһёбһғбҹ’бһӣбһ¶бҹҶбһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһҹбһјбһҳбһ–бҹ’бһҷбһ¶бһҷбһ¶бһҳбһ”бҹ’бһҡбһҫ 3бҹ–


```r
set.seed(2056)
# Kmeans clustering for 3 clusters
kclust <- kmeans(
  df_numeric_select,
  # Specify the number of clusters
  centers = 3,
  # How many random initial configurations
  nstart = 25
)

# Display clustering object
kclust
```

бһңбһҸбҹ’бһҗбһ» kmeans бһҳбһ¶бһ“бһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһ”бһ¶бһ“бһ–бһ·бһ–бһҺбҹҢбһ“бһ¶бһҷбҹүбһ¶бһ„бһӣбһҳбҹ’бһўбһ·бһҸбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„ `help("kmeans()")`бҹ” бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ–бҹҒбһӣбһ“бҹҒбҹҮ бһҷбһҫбһ„бһ“бһ№бһ„бһ•бҹ’бһҸбҹ„бһҸбһӣбһҫбһҒбҹ’бһӣбҹҮбҹ—бһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮбҹ” бһҷбһҫбһ„бһғбһҫбһүбһҗбһ¶ бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҖбҹ’бһҡбһ»бһҳ into 3 бһҖбҹ’бһҡбһ»бһҳбһҳбһ¶бһ“бһ‘бҹҶбһ бҹҶ 65, 110, 111бҹ” бһӣбһ‘бҹ’бһ’бһ•бһӣбһ“бҹ„бҹҮбһҖбҹҸбһҳбһ¶бһ“бһҳбһҮбҹ’бһҲбҹҶбһҠбҹ’бһӢбһ¶бһ“бһҖбҹ’бһҡбһ»бһҳ (means) бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбҹ’бһҡбһ»бһҳбһ‘бһ¶бҹҶбһ„ 3 бһ“бҹ…бһӣбһҫбһўбһҗбҹҒбһҡбһ…бҹҶбһ“бһҪбһ“ 5 бҹ”

бһңбҹүбһ·бһ…бһ‘бҹҗбһҡбһҖбҹ’бһҡбһ»бһҳбһӮбһәбһҮбһ¶бһ…бҹҶбһҺбһ¶бһҸбҹӢбһҗбҹ’бһ“бһ¶бһҖбҹӢбһҖбҹ’бһҡбһ»бһҳбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһҹбһ„бҹ’бһҖбҹҒбһҸбһ“бһёбһҳбһҪбһҷбҹ—бҹ” бһҷбһҫбһ„бһ…бһјбһҡбһ”бҹ’бһҡбһҫбһҳбһ»бһҒбһ„бһ¶бһҡ `augment` бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ“бҹ’бһҗбҹӮбһҳбһ…бҹҶбһҺбһ¶бһҸбҹӢбһҗбҹ’бһ“бһ¶бһҖбҹӢбһҖбҹ’бһҡбһ»бһҳбһ‘бҹ…бһҖбһ¶бһ“бҹӢбһҖбҹҶбһҺбһҸбҹӢбһҸбҹ’бһҡбһ¶бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбһҫбһҳбҹ”


```r
# Add predicted cluster assignment to data set
augment(kclust, df_numeric_select) %>% 
  relocate(.cluster) %>% 
  slice_head(n = 10)
```

бһӣбҹ’бһўбһ”бҹҶбһ•бһ»бһҸ, бһҷбһҫбһ„бһ‘бһҫбһ”бһ”бҹҶбһ”бҹӮбһҖбһҖбҹ’бһҡбһ»бһҳбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҡбһ”бһҹбҹӢбһҷбһҫбһ„бһ‘бҹ…бһҮбһ¶бһҖбҹ’бһҡбһ»бһҳбһ…бҹҶбһ“бһҪбһ“ 3 бһҖбҹ’бһҡбһ»бһҳбҹ” бһ бһҫбһҷбһ бһҫбһҷ, бһҖбһ¶бһҡбһҖбҹ’бһҡбһ»бһҳбһ бҹҠбһ»бһ“бһҡбһ”бһҹбҹӢбһҷбһҫбһ„бһӣбҹ’бһўбһ”бҹүбһ»бһ“бһҺбһ¶ рҹӨ·? бһҷбһҫбһ„бһҳбһҫбһӣбһ‘бҹ…бһӣбһҫ `бһ–бһ·бһ“бҹ’бһ‘бһ» Silhouette`

### **бһ–бһ·бһ“бҹ’бһ‘бһ» Silhouette**

[бһҖбһ¶бһҡвҖӢбһңбһ·бһ—бһ¶бһӮ Silhouette](https://en.wikipedia.org/wiki/Silhouette_(clustering)) бһўбһ¶бһ…бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бҹ’бһҡбһҫбһҠбһҫбһҳбҹ’бһ”бһёбһҹбһ·бһҖбҹ’бһҹбһ¶бһ…бһҳбҹ’бһ„бһ¶бһҷбһ”бҹҶбһ”бҹӮбһҖбһҡбһңбһ¶бһ„бһҖбҹ’бһҡбһ»бһҳбһҠбҹӮбһӣбһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бҹ” бһ–бһ·бһ“бҹ’бһ‘бһ»бһ“бҹҒбҹҮбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡбһ–бһё -1 бһ‘бҹ… 1 бһ бһҫбһҷбһ”бһҫбһ–бһ·бһ“бҹ’бһ‘бһ»бһҮбһ·бһҸ 1 бһ“бҹ„бҹҮбһҖбҹ’бһҡбһ»бһҳбһ“бҹ„бҹҮбһҳбһ¶бһ“бһ—бһ¶бһ–бһҹбһҳбҹ’бһ”бһјбһҡбһ”бҹӮбһ” бһ“бһ·бһ„бһ”бҹҶбһ”бҹӮбһҖбһ…бҹҒбһүбһ–бһёбһҖбҹ’бһҡбһ»бһҳбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһҷбҹүбһ¶бһ„бһ…бҹ’бһ”бһ¶бһҹбҹӢбҹ” бһҸбһҳбҹ’бһӣбҹғбһҮбһ·бһҸ 0 бһҮбһ¶бһҸбҹҶбһҺбһ¶бһ„бһұбҹ’бһҷбһҖбҹ’бһҡбһ»бһҳбһҲбҹ’бһӣбҹ„бҹҮбһӮбҹ’бһ“бһ¶бһҳбһ¶бһ“бһӮбҹҶбһҡбһјбһҠбҹӮбһӣбһ“бҹ…бһҮбһ·бһҸбһҹбҹ’бһ“бһ·бһ‘бҹ’бһ’бһҠбһӣбҹӢбһҹбҹ’бһҡбһҳбҹ„бһӣбһҹбһҳбҹ’бһҡбҹҒбһ…бһ…бһ·бһҸбҹ’бһҸбһҡбһңбһ¶бһ„бһҖбҹ’бһҡбһ»бһҳбһҮбһ·бһҸбһҒбһ¶бһ„бҹ”[бһ”бҹ’бһҡбһ—бһ–](https://dzone.com/articles/kmeans-silhouette-score-explained-with-python-exam)бҹ”

бһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҡбҹ’бһҸ silhouette бһҳбһ’бҹ’бһҷбһҳбһӮбһ·бһҸбһ–бһ·бһ“бҹ’бһ‘бһ» silhouette бһҳбһ’бҹ’бһҷбһҳбһ“бҹғбһҖбһ¶бһҡбһҹбһ„бҹ’бһҖбҹҒбһҸбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҸбһҳбҹ’бһӣбҹғ *k* бһ•бҹ’бһҹбҹҒбһ„бҹ—бһӮбҹ’бһ“бһ¶бҹ” бһ–бһ·бһ“бҹ’бһ‘бһ» silhouette бһҳбһ’бҹ’бһҷбһҳбһҒбҹ’бһ–бһҹбҹӢбһ”бһ„бҹ’бһ бһ¶бһүбһ–бһёбһҖбһ¶бһҡбһ”бҹӮбһ„бһ…бҹӮбһҖбһҖбҹ’бһҡбһ»бһҳбһӣбҹ’бһўбҹ”

бһҳбһ»бһҒбһ„бһ¶бһҡ `silhouette` бһҖбҹ’бһ“бһ»бһ„бһҖбһүбҹ’бһ…бһ”бҹӢ cluster бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бҹ’бһҡбһҫбһҠбһҫбһҳбҹ’бһ”бһёбһӮбһҺбһ“бһ¶бһӣбҹ’бһ”бҹҮ silhouette бһҳбһ’бҹ’бһҷбһҳбҹ”

> silhouette бһўбһ¶бһ…бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһӮбһҺаёҷаёІаё”а№үаё§аёўбһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҡбҹ’бһҸ [бһ…бһҳбҹ’бһ„бһ¶бһҷ](https://en.wikipedia.org/wiki/Distance "бһ…бһҳбҹ’бһ„бһ¶бһҷ") бһ•бҹ’бһҹбҹҒбһ„бҹ—бһҠбһјбһ…бһҮбһ¶ [бһ…бһҳбҹ’бһ„бһ¶бһҷ Euclidean](https://en.wikipedia.org/wiki/Euclidean_distance "бһ…бһҳбҹ’бһ„бһ¶бһҷ Euclidean") бһ¬ [бһ…бһҳбҹ’бһ„бһ¶бһҷ Manhattan](https://en.wikipedia.org/wiki/Manhattan_distance "бһ…бһҳбҹ’бһ„бһ¶бһҷ Manhattan") бһҠбҹӮбһӣбһҷбһҫбһ„бһ”бһ¶бһ“бһ–бһ·бһ—бһ¶бһҖбҹ’бһҹбһ¶бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„ [бһҳбҹҒбһҡбҹҖбһ“бһҳбһ»бһ“](https://github.com/microsoft/ML-For-Beginners/blob/main/5-Clustering/1-Visualize/solution/R/lesson_14-R.ipynb)бҹ”


```r
# Load cluster package
library(cluster)

# Compute average silhouette score
ss <- silhouette(kclust$cluster,
                 # Compute euclidean distance
                 dist = dist(df_numeric_select))
mean(ss[, 3])
```

бһ–бһ·бһ“бҹ’бһ‘бһ»бһҡбһ”бһҹбҹӢбһҷбһҫбһ„бһӮбһә **.549**, бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһӮбһәбһ“бҹ…бһҖбһҺбҹ’бһҸбһ¶бһӣбҹ” бһ“бҹҒбҹҮбһ”бһ„бҹ’бһ бһ¶бһүбһҗбһ¶бһҠбһ¶бһҸбһ¶бһҡбһ”бһҹбҹӢбһҷбһҫбһ„бһҳбһ·бһ“бһҹбһҳбһҹбҹ’бһҡбһ”бһҮбһ¶бһ–бһ·бһҹбҹҒбһҹбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бҹ’бһҡбһ—бҹҒбһ‘ clustering бһ“бҹҒбҹҮбһ‘бҹҒбҹ” бһҷбһҫбһ„бһҳбһҖбһҳбһҫбһӣбһҗбһ¶бһҸбһҫбһҷбһҫбһ„бһўбһ¶бһ…бһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһҹбһ„бҹ’бһҹбҹҗбһҷбһ“бҹҒбҹҮбһҸбһ¶бһҳбһҡбһјбһ”бһҳбһ“бҹ’бһҸбһңбһ·бһҮбҹ’бһҮбһ¶бһҳбһ¶бһ“бһ‘бҹҒбһ¬бһ“бҹ…бҹ” бһҖбһүбҹ’бһ…бһ”бҹӢ [factoextra package](https://rpkgs.datanovia.com/factoextra/index.html) бһ•бҹ’бһҸбһӣбҹӢбһ“бһјбһңбһҳбһ»бһҒбһ„бһ¶бһҡ (`fviz_cluster()`) бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһ бһ¶бһү clusteringбҹ”


```r
library(factoextra)

# Visualize clustering results
fviz_cluster(kclust, df_numeric_select)
```

бһҖбһ¶бһҡбһҷбһҖбһӮбҹ’бһ“бһ¶бһ”бҹүбҹҮбһӮбҹ’бһ“бһ¶бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбҹ’бһҡбһ»бһҳбһ”бһ„бҹ’бһ бһ¶бһүбһҗбһ¶бһҠбһ¶бһҸбһ¶бһҡбһ”бһҹбҹӢбһҷбһҫбһ„бһҳбһ·бһ“бһҹбһҳбһҡбһҳбҹ’бһҷбһ–бһ·бһҹбҹҒбһҹбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бҹ’бһҡбһ—бҹҒбһ‘бһҖбҹ’бһҡбһ»бһҳбһ”бҹӮбһ„бһ…бҹӮбһҖбһ“бҹҒбҹҮбһ‘бҹҒ бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһ‘бҹ…бһҳбһ»бһҒбһ‘бҹ…бҹ”

## 4. бһҖбһ¶бһҡбһҖбҹҶбһҺбһҸбҹӢбһ…бҹҶбһ“бһҪбһ“бһҖбҹ’бһҡбһ»бһҳбһҠбҹӮбһӣбһӣбҹ’бһўбһ”бҹҶбһ•бһ»бһҸ

бһҹбҹҶбһҺбһҪбһҡбһҹбҹҶбһҒбһ¶бһ“бҹӢбһҳбһҪбһҷбһҠбҹӮбһӣбһ“бҹ…бһҸбҹӮбһҳбһ¶бһ“бһҮбһ¶бһүбһ№бһҖбһүбһ¶бһ”бҹӢбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ”бҹӮбһ„бһ…бҹӮбһҖбһҮбһ¶бһҖбҹ’бһҡбһ»бһҳ K-Means бһӮбһәбһ“бҹҒбҹҮ - бһҠбҹ„бһҷбһӮбҹ’бһҳбһ¶бһ“бһҹбҹ’бһӣбһ¶бһҖбһҗбҹҒбһҡбһҮбһ¶бһ”бҹӢбһҗбҹ’бһ“бһ¶бһҖбҹӢбһҠбҹӮбһӣбһ”бһ¶бһ“бһҹбҹ’бһӮбһ¶бһӣбҹӢбһ’бҹ’бһңбһҫбһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…бһҠбһҫбһҳбҹ’бһ”бһёбһҠбһ№бһ„бһ…бҹҶбһ“бһҪбһ“бһҖбҹ’бһҡбһ»бһҳбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһ”бҹҶбһ”бҹӮбһҖбһҠбһ¶бһҸбһ¶бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ?

бһҳбһ’бҹ’бһҷбҹ„бһ”бһ¶бһҷбһҳбһҪбһҷбһҠбҹӮбһӣбһҷбһҫбһ„бһўбһ¶бһ…бһ–бҹ’бһҷбһ¶бһҷбһ¶бһҳбһҡбһҖбһӮбһәбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһҸбҹҶбһҺбһ¶бһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ `бһ”бһ„бҹ’бһҖбһҫбһҸбһҹбҹҠбҹҒбһҡбһёбһ“бҹғбһӮбҹҶбһҡбһјбһ”бҹӮбһ„бһ…бҹӮбһҖбһҖбҹ’бһҡбһ»бһҳ` бһҮбһ¶бһҳбһҪбһҷбһ…бҹҶбһ“бһҪбһ“бһҖбҹ’бһҡбһ»бһҳбһҖбһҫбһ“бһЎбһҫбһ„ (бһ§. бһ–бһё 1-10) бһ бһҫбһҷбһңбһ¶бһҷбһҸбһҳбҹ’бһӣбҹғбһҳбһ¶бһҸбҹ’бһҡбһҠбҹ’бһӢбһ¶бһ“бһ”бҹӮбһ„бһ…бҹӮбһҖбһҖбҹ’бһҡбһ»бһҳбһҠбһјбһ…бһҮбһ¶ **бһ–бһ·бһ“бҹ’бһ‘бһ» Silhouetteбҹ”**

бһҳбһҖбһҖбҹҶбһҺбһҸбҹӢбһ…бҹҶбһ“бһҪбһ“бһҖбҹ’бһҡбһ»бһҳбһӣбҹ’бһўбһ”бҹҶбһ•бһ»бһҸбһҠбҹ„бһҷбһӮбһҺбһ“бһ¶бһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡбһ”бҹӮбһ„бһ…бҹӮбһҖбһҖбҹ’бһҡбһ»бһҳбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҸбһҳбҹ’бһӣбҹғ *k* бһ•бҹ’бһҹбҹҒбһ„бҹ— бһ бһҫбһҷбһңбһ¶бһҷбһҸбһҳбҹ’бһӣбҹғ **бһ•бһӣбһ”бһјбһҖбһҖбҹ’бһ“бһ»бһ„бһ…бҹҶбһҺбҹ„бһҳбһҖбҹ’бһҡбһ»бһҳ (Within Cluster Sum of Squares)** (WCSS)бҹ” бһ•бһӣбһ”бһјбһҖбһҹбһҡбһ»бһ”бһҖбҹ’бһ“бһ»бһ„бһ…бҹҶбһҺбҹ„бһҳбһҖбҹ’бһҡбһ»бһҳ (WCSS) бһңбһ¶бһҹбҹӢбһңбҹӮбһ„бһ—бһ¶бһ–бһҠбһ·бһҸбһҠбҹ’бһҷбһҡбһ”бһҹбҹӢбһҖбһ¶бһҡбһ”бҹӮбһ„бһ…бҹӮбһҖбһҖбҹ’бһҡбһ»бһҳ бһ бһҫбһҷбһҷбһҫбһ„бһ…бһ„бҹӢбһІбҹ’бһҷбһңбһ¶бһҗбһҷбһ…бһ»бҹҮбһ”бҹҶбһ•бһ»бһҸ бһўбҹ„бһҷбһҸбһҳбҹ’бһӣбҹғбһ‘бһ¶бһ”бһҳбһ¶бһ“бһ“бҹҗбһҷбһҗбһ¶бһ…бҹҶбһҺбһ»бһ…бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҹбҹ’бһҗбһ·бһҸбһ“бҹ…бһҮбһ·бһҸбһӮбҹ’бһ“бһ¶бҹ”

бһҳбһҖбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҘбһ‘бҹ’бһ’бһ·бһ–бһӣбһ“бҹғбһҮбһҳбҹ’бһҡбһҫбһҹбһҒбһ»бһҹбҹ—бһӮбҹ’бһ“бһ¶бһҡбһ”бһҹбҹӢ `k` бһ–бһё 1 бһҠбһӣбҹӢ 10 бһ‘бҹ…бһӣбһҫбһҖбһ¶бһҡбһ”бҹӮбһ„бһ…бҹӮбһҖбһҖбҹ’бһҡбһ»бһҳбһ“бҹҒбҹҮбҹ”


```r
# Create a series of clustering models
kclusts <- tibble(k = 1:10) %>% 
  # Perform kmeans clustering for 1,2,3 ... ,10 clusters
  mutate(model = map(k, ~ kmeans(df_numeric_select, centers = .x, nstart = 25)),
  # Farm out clustering metrics eg WCSS
         glanced = map(model, ~ glance(.x))) %>% 
  unnest(cols = glanced)
  

# View clustering rsulsts
kclusts
```

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҠбҹӮбһӣбһҷбһҫбһ„бһҳбһ¶бһ“бһҸбһҳбҹ’бһӣбҹғбһ”бһјбһҖбһҹбһҡбһ»бһ”бһ“бҹғбһ…бҹҶбһ“бһҪбһ“бһ”бһҸбҹӢбһҒбҹ’бһ‘бһ„бҹӢбһҒбһ¶бһ„бһҖбҹ’бһ“бһ»бһ„бһҖбҹ’бһҡбһ»бһҳ (tot.withinss) бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҮAlgorithms бһҖбһ¶бһҡбһ”бҹӮбһ„бһ…бҹӮбһҖбһ“бһёбһҳбһҪбһҷбҹ—бһҠбҹ„бһҷбһҳбһ¶бһ“бһҳбһҮбҹ’бһҲбһҳбһҺбҹ’бһҢбһӣ *k* бһҷбһҫбһ„бһ”бҹ’бһҡбһҫбһңбһ·бһ’бһё [elbow method](https://en.wikipedia.org/wiki/Elbow_method_(clustering)) бһҠбһҫбһҳбҹ’бһ”бһёбһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһ…бҹҶбһ“бһҪбһ“бһҖбҹ’бһҡбһ»бһҳбһҠбҹӮбһӣбһӣбҹ’бһўбһ”бҹҶбһ•бһ»бһҸбҹ” бһңбһ·бһ’бһёбһ“бҹҒбҹҮбһҡбһҪбһҳбһҳбһ¶бһ“бһҖбһ¶бһҡбһҡбҹҖбһ”бһ…бҹҶбһҖбҹ’бһҡбһ¶бһ” WCSS бһҮбһ¶бһўбһ“бһ»бһӮбһҳбһ“бҹҚбһ“бҹғбһ…бҹҶбһ“бһҪбһ“бһҖбҹ’бһҡбһ»бһҳ бһ бһҫбһҷбһҮбҹ’бһҡбһҫбһҹбһҷбһҖ [elbow of the curve](https://en.wikipedia.org/wiki/Elbow_of_the_curve "Elbow of the curve") бһҮбһ¶бһ…бҹҶбһ“бһҪбһ“бһҖбҹ’бһҡбһ»бһҳбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһ”бҹ’бһҡбһҫбҹ”


```r
set.seed(2056)
# Use elbow method to determine optimum number of clusters
kclusts %>% 
  ggplot(mapping = aes(x = k, y = tot.withinss)) +
  geom_line(size = 1.2, alpha = 0.8, color = "#FF7F0EFF") +
  geom_point(size = 2, color = "#FF7F0EFF")
```

бһҡбһјбһ”бһ—бһ¶бһ–бһ”бһ„бҹ’бһ бһ¶бһүбһ–бһёбһҖбһ¶бһҡбһҖбһ¶бһҸбҹӢбһ”бһ“бҹ’бһҗбһҷ WCSS бһҹбҹҶбһҒбһ¶бһ“бҹӢбһҳбһҪбһҷ (бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮвҖӢбһҳбһ¶бһ“ *бһ—бһ¶бһ–бһҸбһ№бһ„*) бһҒбһҺбҹҲбһҠбҹӮбһӣбһ…бҹҶбһ“бһҪбһ“бһҖбҹ’бһӣбһ»бһҹбҹ’бһ‘бһәбһҡбһҖбһҫбһ“бһ–бһёбһҳбһҪбһҷбһ‘бҹ…бһ–бһёбһҡ бһ бһҫбһҷбһҳбһ¶бһ“бһҖбһ¶бһҡбһҖбһ¶бһҸбҹӢбһ”бһ“бҹ’бһҗбһҷбһ…бҹ’бһ”бһ¶бһҹбҹӢбһӣбһ¶бһҹбҹӢбһ”бһ“бҹ’бһҗбҹӮбһҳбһ‘бҹҖбһҸбһ–бһёбһ–бһёбһҡбһ‘бҹ…бһ”бһёбһҖбҹ’бһӣбһ»бһҹбҹ’бһ‘бһәбһҡбҹ” бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖ бһҖбһ¶бһҡбһҖбһ¶бһҸбҹӢбһ”бһ“бҹ’бһҗбһҷбһҖбһ¶бһ“бҹӢбһҸбҹӮбһҸбһ·бһ…бһ”бһ“бҹ’бһҗбһҷ бһўбҹ„бһҷбһҳбһ¶бһ“ `elbow` рҹ’Ә бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҸбһ¶бһҡбһ¶бһ„бһ“бҹ…бһҮбһ·бһҸбһ”бһёбһҖбҹ’бһӣбһ»бһҹбҹ’бһ‘бһәбһҡбҹ” бһ“бҹҒбҹҮбһҮбһ¶бһҹбһүбҹ’бһүбһ¶бһӣбҹ’бһўбһҗбһ¶бһҳбһ¶бһ“бһ–бһёбһҡбһҠбһӣбҹӢбһ”бһёбһҖбҹ’бһӣбһ»бһҹбҹ’бһ‘бһәбһҡбһҠбҹӮбһӣбһ”бҹҶбһ”бҹӮбһҖбһ”бһ¶бһ“бһӣбҹ’бһўбһ“бҹғбһ…бҹҶбһҺбһ»бһ…бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбҹ”

бһ–бҹҒбһӣбһ“бҹҒбҹҮбһҷбһҫбһ„бһўбһ¶бһ…бһ”бһ“бҹ’бһҸбһ‘бҹ…бһҷбһҖбһҳбҹүбһјбһҠбҹӮбһӣбһҖбҹ’бһӣбһ»бһҹбҹ’бһ‘бһәбһҡбһҠбҹӮбһӣ `k = 3` бҹ–

> `pull()`: бһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҷбһҖбһҮбһҪбһҡбһҲбһҡбһҸбҹӮбһҳбһҪбһҷ  
>
> `pluck()`: бһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһүбҹ’бһҮбһёбһ¬бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбһјбһ…бһҮбһ¶ бһ”бһүбҹ’бһҮбһё


```r
# Extract k = 3 clustering
final_kmeans <- kclusts %>% 
  filter(k == 3) %>% 
  pull(model) %>% 
  pluck(1)


final_kmeans
```

бһҹбһјбһҳбһўбһҡбһӮбһ»бһҺ! бһІбҹ’бһҷбһҷбһҫбһ„бһ”бһ“бҹ’бһҸбһ”бһ„бҹ’бһ бһ¶бһү cluster бһҠбҹӮбһӣбһ”бһ¶бһ“бһ‘бһ‘бһҪбһӣбҹ” бһҸбһҫбһ…бһ„бҹӢбһҳбһ¶бһ“бһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһҳбһҪбһҷбһ”бҹ’бһҡбһҫ `plotly` бһҠбҹӮбһҡбһ¬бһ‘бҹҒ?


```r
# Add predicted cluster assignment to data set
results <-  augment(final_kmeans, df_numeric_select) %>% 
  bind_cols(df_numeric %>% select(artist_top_genre)) 

# Plot cluster assignments
clust_plt <- results %>% 
  ggplot(mapping = aes(x = popularity, y = danceability, color = .cluster, shape = artist_top_genre)) +
  geom_point(size = 2, alpha = 0.8) +
  paletteer::scale_color_paletteer_d("ggthemes::Tableau_10")

ggplotly(clust_plt)
```

бһ”бҹ’бһҡбһ бҹӮбһӣбһҮбһ¶бһҷбһҫбһ„бһ”бһ¶бһ“бһҡбҹҶбһ–бһ№бһ„бһҗбһ¶clusterбһ“бһёбһҳбһҪбһҷбҹ—(бһҠбҹӮбһӣбһҸбҹҶбһҺбһ¶бһ„бһҠбҹ„бһҷбһ–бһҺбҹҢбһ•бҹ’бһҹбҹҒбһ„бҹ—)бһ“бһ№бһ„бһҳбһ¶бһ“бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ—бҹӮбһ“бһҒбһ»бһҹбһӮбҹ’бһ“бһ¶(бһҠбҹӮбһӣбһҸбҹҶбһҺбһ¶бһ„бһҠбҹ„бһҷбһҡбһ¶бһ„бһ’бһ¶бһҸбһ»бһ•бҹ’бһҹбҹҒбһ„бҹ—)бҹ”

бһҷбһҫбһ„бһҳбһҖбһҹбһ¶бһҖбһҹбһҪбһҡбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһ—бһ¶бһ–бһҸбҹ’бһҡбһјбһңбһҸбһ¶бһҳбһҳбҹүбһјбһҠбҹӮбһӣбҹ”


```r
# Assign genres to predefined integers
label_count <- results %>% 
  group_by(artist_top_genre) %>% 
  mutate(id = cur_group_id()) %>% 
  ungroup() %>% 
  summarise(correct_labels = sum(.cluster == id))


# Print results  
cat("Result:", label_count$correct_labels, "out of", nrow(results), "samples were correctly labeled.")

cat("\nAccuracy score:", label_count$correct_labels/nrow(results))
```

-   бһҳбһҫбһӣбһ§бһ”бһҖбһҡбһҺбҹҚбһҹбһҳбҹ’бһҡбһ„бҹӢбһҖбҹ’бһҡбһ»бһҳ K-Means [бһҠбһјбһ…бһҳбһҪбһҷбһ“бҹҒбҹҮ](https://user.ceng.metu.edu.tr/~akifakkus/courses/ceng574/k-means/). бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ§бһ”бһҖбһҡбһҺбҹҚбһ“бҹҒбҹҮбһҠбһҫбһҳбҹ’бһ”бһёбһҳбһҫбһӣбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһ“бһ·бһ„бһҖбҹҶбһҺбһҸбҹӢбһўбҹҶбһ–бҹӮбһҖбһҳбҹ’бһ–бһ»бһҮбһ¶ (centroids) бһҡбһ”бһҹбҹӢбһңбһ¶бҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҖбҹӮбһ”бҹ’бһҡбҹӮбһ—бһ¶бһ–бһ…бҹғбһҠбһ“бҹ’бһҷбһ“бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ бһ…бҹҶбһ“бһҪбһ“бһҖбҹ’бһҡбһ»бһҳ бһ“бһ·бһ„бһ…бҹҶбһ“бһҪбһ“бһўбҹҶбһ–бҹӮбһҖбһҳбҹ’бһ–бһ»бһҮбһ¶бҹ” бһңбһ¶бһҳбһ¶бһ“бһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһҷбһӣбҹӢбһ–бһёбһҡбһ”бҹҖбһ”бһҠбҹӮбһӣбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһўбһ¶бһ…бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ—бҹ’бһҮбһ¶бһ”бҹӢбһҖбҹ’бһҡбһ»бһҳбһўбҹ’бһңбһёбһҳбһҪбһҷ?

-   бһ•бһ„бһҠбҹӮбһҡ бһҳбһҫбһӣбһҜбһҖбһҹбһ¶бһҡбһ“бҹҒбҹҮ [бһўбҹҶбһ–бһё K-Means](https://stanford.edu/~cpiech/cs221/handouts/kmeans.html) бһ–бһёбһҹбһ¶бһҖбһӣбһңбһ·бһ‘бҹ’бһҷбһ¶бһӣбҹҗбһҷ Stanfordбҹ”

бһ…бһ„бҹӢбһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһҮбҹҶбһ“бһ¶бһүбһ”бҹӮбһ„бһ…бҹӮбһҖбһҖбҹ’бһҡбһ»бһҳбһҗбҹ’бһҳбһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ‘бҹ…бһҮбһ¶бһҹбҹҶбһҺбһ»бҹҶбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹӮбһӣбһҹбһҳбһҹбҹ’бһҡбһ”бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢ K-Means бһҠбҹӮбһҡбһ¬бһ‘бҹҒ? бһҹбһјбһҳбһҳбһҫбһӣбҹ–

-   [бһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣ бһ“бһ·бһ„бһңбһ¶бһҷбһҸбһҳбҹ’бһӣбҹғбһҳбҹүбһјбһҠбҹӮбһӣбһ”бҹӮбһ„бһ…бҹӮбһҖбһҖбҹ’бһҡбһ»бһҳ](https://rpubs.com/eR_ic/clustering) бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ Tidymodels бһ“бһ·бһ„бһҳбһ·бһҸбҹ’бһҸбһ—бһҖбҹ’бһҸбһ·

-   [бһңбһ·бһ—бһ¶бһӮбһҖбҹ’бһҡбһ»бһҳ K-means](https://uc-r.github.io/kmeans_clustering), бһӮбһҺбһ“бһёбһңбһ·бһ‘бҹ’бһҷбһ¶бһҮбҹҶбһ“бһ¶бһүбһҡбһ”бһҹбҹӢ UC бһ•бһӣбһ·бһҸбһҖбһҳбҹ’бһҳбһўбһ¶бһҮбһёбһңбһҖбһҳбҹ’бһҳ

- [бһҖбһ¶бһҡбһ”бҹӮбһ„бһ…бҹӮбһҖбһҖбҹ’бһҡбһ»бһҳ K-means бһҮбһ¶бһҳбһҪбһҷбһӮбҹ„бһӣбһҖбһ¶бһҡбһҺбҹҚбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ tidy](https://www.tidymodels.org/learn/statistics/k-means/)
