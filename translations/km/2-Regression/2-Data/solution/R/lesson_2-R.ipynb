бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“:

- бһҡбһ”бҹҖбһ”бһҡбҹҖбһ”бһ…бҹҶбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҳбҹүбһјбһҠбҹӮбһӣбҹ”

- бһҡбһ”бҹҖбһ”бһ”бҹ’бһҡбһҫ `ggplot2` бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһҳбһҫбһӣбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбҹ”

бһҹбҹҶбһҺбһҪбһҡбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бҹҶбһ•бһ»бһҸбһ“бһ№бһ„бһҖбҹҶбһҺбһҸбҹӢбһ”бҹ’бһҡбһ—бҹҒбһ‘бһ“бҹғбһўбһ¶бһӣбҹӢбһ бҹ’бһӮбһҡбһёбһ’бһҳбҹҢ ML бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбҹ” бһ бһҫбһҷбһӮбһ»бһҺбһ—бһ¶бһ–бһ“бҹғбһ…бһҳбҹ’бһӣбһҫбһҷбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бһ“бһ№бһ„бһ–бһ№бһ„бһ•бҹ’бһўбҹӮбһҖбһҷбҹүбһ¶бһ„бһҒбҹ’бһӣбһ¶бҹҶбһ„бһӣбһҫбһ’бһҳбҹ’бһҳбһҹбһҳбҹ’бһ”бһҸбҹ’бһҸбһ·бһ“бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбҹ”

## 1. бһ“бһ¶бҹҶбһ…бһјбһӣбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ•бһӣбһӣбҹ’бһ бһ»бһ„ бһ“бһ·бһ„бһ бҹ…бһҳбһҖ Tidyverse

бһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбһҠбһҫбһҳбҹ’бһ”бһёбһҖбҹҶбһҺбһҸбҹӢ бһ“бһ·бһ„бһҡбһ¶бһҷбһ…бҹҒбһүбһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮбҹ–

-   `tidyverse`: [tidyverse](https://www.tidyverse.org/) бһӮбһәбһҮбһ¶ [бһҖбҹҶбһҺбһҸбҹӢбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ R](https://www.tidyverse.org/packages) бһҠбҹӮбһӣбһҡбһ…бһ“бһ¶бһЎбһҫбһ„бһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫбһұбҹ’бһҷбһңбһ·бһ‘бҹ’бһҷбһ¶бһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҡбһ бҹҗбһҹ бһҡбһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣ бһ“бһ·бһ„бһҖбҹ’бһҠбҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһӣбҹҒбһ„бһҮбһ¶бһ„бһҳбһ»бһ“!

бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҠбҹҶбһЎбһҫбһ„бһңбһ¶бһҸбһ¶бһҳбһҡбһҷбҹҲбҹ–

`install.packages(c("tidyverse"))`

бһҹбҹ’бһӮбҹ’бһҡбһёбһ”бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҳбһҫбһӣбһҗбһ¶бһҸбһҫбһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһүбҹ’бһ…бһ”бҹӢбһҳбһјбһҢбһ»бһӣбһ“бҹҒбҹҮбһ бһҫбһҷбһҠбҹҶбһЎбһҫбһ„бһңбһ¶бһҠбһҫбһҳбҹ’бһ”бһёбһҮбҹҶбһ“бһҪбһҹбһ”бһҫбһҳбһ¶бһ“бһҒбһҖбһҒбһ¶бһ“бҹ”


```r
suppressWarnings(if(!require("pacman")) install.packages("pacman"))
pacman::p_load(tidyverse)
```

бһҘбһЎбһјбһңбһ“бҹҒбҹҮ бһҷбһҫбһ„бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһ”бҹӢбһҳбһҪбһҷбһ…бҹҶбһ“бһҪбһ“ бһ“бһ·бһ„бһ•бҹ’бһ‘бһ»бһҖ [бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ](https://github.com/microsoft/ML-For-Beginners/blob/main/2-Regression/data/US-pumpkins.csv) бһҠбҹӮбһӣбһ”бһ¶бһ“бһ•бҹ’бһҸбһӣбҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ!


```r
# Load the core Tidyverse packages
library(tidyverse)

# Import the pumpkins data
pumpkins <- read_csv(file = "https://raw.githubusercontent.com/microsoft/ML-For-Beginners/main/2-Regression/data/US-pumpkins.csv")


# Get a glimpse and dimensions of the data
glimpse(pumpkins)


# Print the first 50 rows of the data set
pumpkins %>% 
  slice_head(n =50)
```

бһҳбһҫбһӣ `glimpse()` бһ–бҹҒбһӣбһңбҹҒбһӣбһ¶вҖӢбһҳбһҪбһҷбһ—бҹ’бһӣбһ¶бһҳ бһҮбһ¶бһ”бһ“бҹ’бһ‘бһ¶бһ“бҹӢбһ”бһ„бҹ’бһ бһ¶бһүбһҗбһ¶вҖӢбһҳбһ¶бһ“бһ…бһ“бҹ’бһӣбҹ„бҹҮбһ‘бһ‘бҹҒ бһ“бһ·бһ„бһҳбһ¶бһ“бһҖбһ¶бһҡбһӣбһ¶бһҷбһ”бһүбҹ’бһ…бһјбһӣбһӮбҹ’бһ“бһ¶бһҡбһңбһ¶бһ„бһҒбҹ’бһҹбҹӮбһўбһҖбҹ’бһҹбһҡ (`chr`) бһ“бһ·бһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһӣбҹҒбһҒ (`dbl`)бҹ” `Date` бһӮбһәбһҮбһ¶бһ”бҹ’бһҡбһ—бҹҒбһ‘бһҒбҹ’бһҹбҹӮбһўбһҖбҹ’бһҹбһҡ бһ бһҫбһҷбһҳбһ¶бһ“бһҖбһјбһЎбһ»бҹҶбһҺбҹҚбһҳбһҪбһҷбһҠбҹҸбһўбһҹбҹ’бһ…бһ¶бһҡбҹ’бһҷбһ бҹ…бһҗбһ¶ `Package` бһҠбҹӮбһӣбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ“бҹ„бҹҮбһҳбһ¶бһ“бһҖбһ¶бһҡбһӣбһ¶бһҷбһ”бһүбҹ’бһ…бһјбһӣбһӮбҹ’бһ“бһ¶бһҡбһңбһ¶бһ„ `sacks`, `bins` бһ“бһ·бһ„бһҸбһҳбҹ’бһӣбҹғбһ•бҹ’бһҹбҹҒбһ„бҹ—бһ‘бҹҖбһҸбҹ” бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ“бҹҒбҹҮ бһҮбһ¶бһҖбһ¶бһҡбһўбһ—бһ·бһңбһҢбҹ’бһҚбһ“бҹҚбһҳбһҪбһҷбһ–бҹ’бһҡбһ¶вҖӢбһҮбһ¶бһҖбҹӢбһҹбҹ’бһҸбҹӮбһ„ рҹҳӨбҹ”

бһ–бһ·бһҸбһ”бҹ’бһҡбһ¶бһҖбһҠ бһңбһ¶бһ“бҹ…бһҳбһ·бһ“бһ’бһҳбҹ’бһҳбһҸбһ¶бһ‘бҹҒбһҠбҹӮбһӣбһ“бһ№бһ„бһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бһҜбһҖбһҹбһ¶бһҡбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ бһҠбҹӮбһӣбһ”бһ¶бһ“бһҡбҹҖбһ”бһ…бҹҶбһҡбһҪбһ…бһҡбһ¶бһӣбҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бҹ’бһҡбһҫбһ”бһ„бҹ’бһҖбһҫбһҸбһҳбҹүбһјбһҢбҹӮбһӣ ML бһ–бһёбһ”бҹ’бһҡбһўбһ”бҹӢбһ‘бҹҒбҹ” бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһҖбһ»бҹҶбһ”бһ¶бһҡбһҳбҹ’бһ— бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“бһ–бһёбһҡбһ”бҹҖбһ”бһҡбҹҖбһ”бһ…бҹҶбһҜбһҖбһҹбһ¶бһҡбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбһҫбһҳбһІбҹ’бһҷбһ”бһ¶бһ“бһҸбҹ’бһҡбһ№бһҳбһҸбҹ’бһҡбһјбһңбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ R бһҠбҹӮбһӣбһҳбһ¶бһ“бһҹбҹ’бһҸбһ„бҹӢбһҠбһ¶бһҡ рҹ§‘вҖҚрҹ”§бҹ” бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“бһ–бһёбһ”бһ…бҹ’бһ…бҹҒбһҖбһ‘бҹҒбһҹбһ•бҹ’бһҹбҹҒбһ„бҹ—бһҠбһҫбһҳбҹ’бһ”бһёбһҹбһҳбҹ’бһҡбһ”бһҹбһҳбҹ’бһҡбһҪбһӣбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ•бһ„бһҠбҹӮбһҡ рҹ“Ҳрҹ“Ҡ

> бһҖбһ¶бһҡбһҡбҹҶбһӣбһ№бһҖбһҳбҹ’бһҠбһ„бһ‘бҹҖбһҸбҹ– бһўбһјбһ”бғ”бғ бғҗбғўбһә pipe (`%>%`) бһ”бҹ’бһҡбһҸбҹ’бһҸбһ·бһ”бһҸбҹ’бһҠбһ·бһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷбһҖбҹҶбһҺбһҸбҹӢбһҸбҹ’бһҡбһ¶бһ“бҹ…бһӣбҹҶбһҠбһ¶бһ”бҹӢбһҸбһ¶бһҳбһ бҹҒбһҸбһ»бһ•бһӣ бһҠбҹ„бһҷбһ•бҹ’бһ‘бҹҒбһҡбһ”бҹ’бһҡбһ’бһ¶бһ“бһңбһҸбҹ’бһҗбһ»бһҳбһҪбһҷбһ‘бҹ…бһҳбһ»бһҒбһ‘бҹ…бһҳбһ»бһҒбһҖбҹ’бһ“бһ»бһ„бһҹбһҳбһёбһҖбһ¶бһҡбһ¬бһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһӮбһ·бһҸбһ–бһёбһўбһјбһ”бғ”бғ бғҗбғўбһә pipe бһҠбһјбһ…бһҮбһ¶бһҖбһ¶бһҡбһ“бһ·бһҷбһ¶бһҷбһҗбһ¶ "бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖ" бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһјбһҠбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбҹ”


## 2. бһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһғбһҫбһүбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҒбҹ’бһңбҹҮ

бһҳбһҪбһҷбһҖбҹ’бһ“бһ»бһ„бһ…бҹҶбһҺбҹ„бһҳбһ”бһүбҹ’бһ бһ¶бһ’бһҳбҹ’бһҳбһҸбһ¶бһ”бҹҶбһ•бһ»бһҸбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһңбһ·бһ‘бҹ’бһҷбһ¶бһҹбһ¶бһҹбҹ’бһҡбҹ’бһҸбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҸбҹ’бһҡбһјбһңбһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһӮбһәбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҳбһ·бһ“бһ–бҹҒбһүбһӣбҹҒбһүбһ¬бһҒбҹ’бһңбҹҮбҹ” R бһҸбҹҶбһҺбһ¶бһ„бһІбҹ’бһҷбһҸбһҳбҹ’бһӣбҹғбһҒбҹ’бһңбҹҮ бһ¬бһҳбһ·бһ“бһҠбһ№бһ„бһҮбһ¶бһҳбһҪбһҷбһҸбһҳбҹ’бһӣбҹғбһҹбһүбҹ’бһүбһ¶бһ–бһ·бһҹбҹҒбһҹбҹ– `NA` (бһҳбһ·бһ“бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ”бһ¶бһ“)бҹ”

бһҠбһјбһ…бһ“бҹҒбҹҮ бһҸбһҫбһҷбһҫбһ„бһ’бҹ’бһңбһҫбһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…бһҠбһҫбһҳбҹ’бһ”бһёбһҠбһ№бһ„бһҗбһ¶ data frame бһҳбһ¶бһ“бһҸбһҳбҹ’бһӣбҹғбһҒбҹ’бһңбҹҮ?
<br>
-   бһңбһ·бһ’бһёбһҳбһҪбһҷбһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҹбһ¶бһҳбһүбҹ’бһүбһӮбһәбһ”бҹ’бһҡбһҫбһҳбһ»бһҒбһ„бһ¶бһҡ base R `anyNA` бһҠбҹӮбһӣбһҸбҹ’бһҡбһЎбһ”бҹӢбһҸбһҳбҹ’бһӣбҹғбһҸбһ»бһӣбҹ’бһҷбһ—бһ¶бһ– `TRUE` бһ¬ `FALSE`


```r
pumpkins %>% 
  anyNA()
```

бһўбһҹбҹ’бһ…бһ¶бһҡбҹ’бһҷбһҺбһ¶бһҹбҹӢ бһҳбһ¶бһ“бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҒбҹ’бһңбҹҮбһҒбһ¶бһҸбһҒбҹ’бһӣбҹҮ! бһ“бҹҒбҹҮбһҮбһ¶бһ‘бһёбһҖбһ“бҹ’бһӣбҹӮбһ„бһӣбҹ’бһўбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ…бһ¶бһ”бҹӢбһ•бҹ’бһҠбһҫбһҳбҹ”

-   бһңбһ·бһ’бһёбһҳбһҪбһҷбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһӮбһәбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһҳбһ»бһҒбһ„бһ¶бһҡ `is.na()` бһҠбҹӮбһӣбһ”бһ„бҹ’бһ бһ¶бһүбһ–бһёбһ’бһ¶бһҸбһ»бһ‘бһ¶бҹҶбһ„бһҳбһјбһӣбһҖбҹ’бһ“бһ»бһ„бһҮбһҪбһҡбһҲбһҡбһҠбҹӮбһӣбһҒбҹ’бһңбҹҮбһҮбһ¶бһҳбһҪбһҷбһҸбһҳбҹ’бһӣбҹғбһҸбҹ’бһҡбһ№бһҳбһҸбҹ’бһҡбһјбһң `TRUE`бҹ”


```r
pumpkins %>% 
  is.na() %>% 
  head(n = 7)
```

бһ”бһ¶бһ“бһ бһҫбһҷ бһ”бһ¶бһ“бһ”бҹҶбһ–бҹҒбһүбһҖбһ¶бһҡбһ„бһ¶бһҡбһҡбһҪбһ… бһҸбҹӮбһ”бһҫбһҮбһ¶бһҳбһҪбһҷвҖӢ data frame бһ’бҹҶбһҠбһјбһ…бһҳбһҪбһҷбһ“бҹҒбҹҮ бһңбһ¶бһ“бһ№бһ„бһўбһҹбҹ’бһҗбһ·бһҡбһ—бһ¶бһ– бһ бһҫбһҷбһҮбһ¶бһҖбҹӢбһҹбҹ’бһҸбҹӮбһ„бһҳбһ·бһ“бһўбһ¶бһ…бһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҳбһҫбһӣбһҮбһҪбһҡбһҲбһҡбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ”бҹ’бһҡбһўбһ”бҹӢбһ‘бһ¶бҹҶбһ„бһўбһҹбҹӢбһ”бһ¶бһ“бһ“бҹ„бҹҮбһ‘бҹҒрҹҳҙбҹ”

-   бһңбһ·бһ’бһёбһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһҮбһ¶бһ„бһӮбҹҒбһ‘бҹҖбһҸбһӮбһәбһӮбһҺбһ“бһ¶бһҠбһӣбҹӢбһ…бҹҶбһ“бһҪбһ“бһҸбһҳбҹ’бһӣбҹғбһҠбҹӮбһӣбһҖбҹҶбһ–бһ»бһ„бһҹбҹ’бһ‘бҹҮбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҮбһҪбһҡбһҲбһҡбһ“бһёбһҳбһҪбһҷбҹ—бҹ–


```r
pumpkins %>% 
  is.na() %>% 
  colSums()
```

бһӣбҹ’бһўбһҮбһ¶бһ„бһҳбһ»бһ“! бһҳбһ¶бһ“бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҒбҹ’бһңбҹҮ бһҒбҹ’бһӣбҹҮбһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһўбһ¶бһ…бһ“бһ№бһ„бһҳбһ·бһ“бһҳбһ¶бһ“бһ”бһүбҹ’бһ бһ¶бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ„бһ¶бһҡбһҠбҹӮбһӣбһҖбҹҶбһ–бһ»бһ„бһ’бҹ’бһңбһҫбһЎбһҫбһҷбҹ” бһҷбһҫбһ„бһҳбһҖбһҳбһҫбһӣбһҗбһ¶бһңбһ·бһ—бһ¶бһӮбһ”бһ“бҹ’бһҗбҹӮбһҳбһ“бһ№бһ„бһ•бҹ’бһҠбһӣбҹӢбһўбҹ’бһңбһёбһҒбҹ’бһӣбҹҮбҹ”

> бһҮбһ¶бһҳбһҪбһҷбһ“бһ№бһ„бһҜбһҖбһҹбҹ’бһҹбһ·бһ‘бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһүбҹ’бһ…бһ”бҹӢ бһ“бһ·бһ„бһҳбһ»бһҒбһ„бһ¶бһҡбһҠбҹҸбһўбһҹбҹ’бһ…бһ¶бһҡбҹ’бһҷ R бһҳбһ¶бһ“бһҜбһҖбһҹбһ¶бһҡбһҠбҹҸбһӣбҹ’бһўбһҺбһ¶бһҹбҹӢбҹ” бһҮбһ¶бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһҹбһјбһҳбһ”бҹ’бһҡбһҫ `help(colSums)` бһ¬ `?colSums` бһҠбһҫбһҳбҹ’бһ”бһёбһҹбҹ’бһңбҹӮбһ„бһҷбһӣбҹӢбһ”бһ“бҹ’бһҗбҹӮбһҳбһўбҹҶбһ–бһёбһҳбһ»бһҒбһ„бһ¶бһҡбҹ”


## 3. Dplyr: бһңбҹҒбһҷбҹ’бһҷбһ¶бһҖбһҡбһҺбҹҚбһ“бҹғбһҖбһ¶бһҡбһ”бҹҶбһӣбҹӮбһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ

[`dplyr`](https://dplyr.tidyverse.org/), бһӮбһәбһҮбһ¶бһ•бҹ’бһ“бҹӮбһҖбһ”бһ“бҹ’бһҗбҹӮбһҳбһҳбһҪбһҷбһҖбҹ’бһ“бһ»бһ„ Tidyverse бһҠбҹӮбһӣбһҮбһ¶бһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҡбҹ’бһҸбһ“бҹғбһҖбһ¶бһҡбһ”бһҳбҹ’бһҡбһҫбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ бһҠбҹӮбһӣбһ•бҹ’бһҸбһӣбҹӢбһ“бһјбһңбһҖбһ·бһҡбһ·бһҷбһ¶бһҹбҹҗбһ–бҹ’бһ‘бһҳбһҪбһҷбһҠбҹӮбһӣбһҳбһ¶бһ“бһҹбһ‘бһ·бҹ’бһ’бһ—бһ¶бһ– бһҠбҹӮбһӣбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһ”бһүбҹ’бһ бһ¶бһҡбһ”бһҹбҹӢбһҖбһ¶бһҡбһ”бһҳбҹ’бһҡбһҫбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹӮбһӣбһҮбһ¶бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ‘бһјбһ‘бҹ…бҹ” бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ•бҹ’бһ“бҹӮбһҖбһ“бҹҒбҹҮ бһҷбһҫбһ„бһ“бһ№бһ„бһҹбҹ’бһңбҹӮбһ„бһҷбһӣбҹӢбһўбҹҶбһ–бһёбһҖбһ·бһҡбһ·бһҷбһ¶бһҹбҹҗбһ–бҹ’бһ‘бһҳбһҪбһҷбһ…бҹҶбһ“бһҪбһ“бһҡбһ”бһҹбҹӢ dplyr!  
<br>


#### dplyr::select()

`select()` бһӮбһәбһҮбһ¶бһ§бһ”бһҖбһҡбһҺбҹҚбһҳбһҪбһҷбһҖбҹ’бһ“бһ»бһ„бһҖбһүбҹ’бһ…бһ”бҹӢ `dplyr` бһҠбҹӮбһӣбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҮбһҪбһҡбһҲбһҡбһҠбҹӮбһӣбһ…бһ„бҹӢбһҡбһҖбҹ’бһҹбһ¶бһ‘бһ»бһҖбһ¬бһҠбһҖбһ…бҹҒбһүбҹ”

бһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫбһІбҹ’бһҷбһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ бһӣбһ”бҹӢбһҮбһҪбһҡбһҲбһҡбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҡбһ”бһҹбҹӢбһңбһ¶бһ…бҹҒбһү бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ `select()` бһҡбһҖбҹ’бһҹбһ¶бһ‘бһ»бһҖбһҸбҹӮбһҮбһҪбһҡбһҲбһҡбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбҹ”

бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһҖбҹ’бһ“бһ»бһ„бһӣбҹҶбһ бһ¶бһҸбҹӢбһ“бҹҒбҹҮ бһҖбһ¶бһҡбһңбһ·бһ—бһ¶бһӮбһҡбһ”бһҹбҹӢбһҷбһҫбһ„бһ“бһ№бһ„бһ–бһ¶бһҖбҹӢбһ–бҹҗбһ“бҹ’бһ’бһ“бһ№бһ„бһҮбһҪбһҡбһҲбһҡ `Package`, `Low Price`, `High Price` бһ“бһ·бһ„ `Date` бҹ” бһўбһ„бҹ’бһӮбһ»бһҷбһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҮбһҪбһҡбһҲбһҡбһ‘бһ¶бҹҶбһ„бһ“бҹҒбҹҮбҹ”


```r
# Select desired columns
pumpkins <- pumpkins %>% 
  select(Package, `Low Price`, `High Price`, Date)


# Print data set
pumpkins %>% 
  slice_head(n = 5)
```

#### dplyr::mutate()

`mutate()` бһӮбһәбһҮбһ¶бһўбһ“бһ»бһӮбһҳбһ“бҹҚбһҳбһҪбһҷбһҖбҹ’бһ“бһ»бһ„бһҖбһүбҹ’бһ…бһ”бҹӢ `dplyr` бһҠбҹӮбһӣбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһ”бһ„бҹ’бһҖбһҫбһҸбһ¬бһҖбҹӮбһ”бҹ’бһҡбҹӮбһҮбһҪбһҡбһҲбһҡ бһҒбһҺбҹҲбһҠбҹӮбһӣбһҡбһҖбҹ’бһҹбһ¶бһҮбһҪбһҡбһҲбһҡбһҠбһҫбһҳбһ‘бҹ…бҹ”

бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһ‘бһјбһ‘бҹ…бһ“бҹғ mutate бһӮбһәбҹ–

`data %>%   mutate(new_column_name = what_it_contains)`

бһҷбһҫбһ„бһҳбһ¶бһ“бһӣбҹ’бһ”бҹӮбһ„ mutate бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҮбһҪбһҡбһҲбһҡ `Date` бһҠбҹ„бһҷбһ’бҹ’бһңбһҫбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

1.  бһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡбһ–бҹҒбһӣбһңбҹҒбһӣбһ¶ (бһҠбҹӮбһӣбһҖбҹҶбһ–бһ»бһ„бһҮбһ¶ бһҸбһҪбһўбһҖбҹ’бһҹбһҡбһ‘бһ‘бҹҒ) бһ‘бҹ…бһҮбһ¶ бһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷбһҒбҹӮ (бһ‘бһ¶бҹҶбһ„бһ“бҹҒбҹҮбһӮбһәбһҮбһ¶бһҖбһ¶бһӣбһ”бһҡбһ·бһ…бҹ’бһҶбҹҒбһ‘бһўбһ¶бһҳбҹҒбһҡбһ·бһҖ бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷбһӮбһә `MM/DD/YYYY`)бҹ”

2.  бһҠбһҖбһҒбҹӮбһ…бҹҒбһүбһ–бһёбһҖбһ¶бһӣбһ”бһҡбһ·бһ…бҹ’бһҶбҹҒбһ‘бһ‘бҹ…бһҮбһ¶вҖӢбһҮбһҪбһҡбһҲбһҡбһҗбҹ’бһҳбһёбһҳбһҪбһҷбҹ”

бһҖбҹ’бһ“бһ»бһ„ R бһҖбһүбҹ’бһ…бһ”бҹӢ [lubridate](https://lubridate.tidyverse.org/) бһ’бҹ’бһңбһҫбһІбҹ’бһҷбһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ–бҹҒбһӣбһңбҹҒбһӣбһ¶бҹ” бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮ бһҳбһҖбһ”бҹ’бһҡбһҫ `dplyr::mutate()`, `lubridate::mdy()`, `lubridate::month()` бһ бһҫбһҷбһҳбһҫбһӣбһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡбһҠбһҫбһҳбҹ’бһ”бһёбһҹбһҳбҹ’бһҡбҹҒбһ…бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҒбһ¶бһ„бһӣбһҫбҹ” бһҷбһҫбһ„бһўбһ¶бһ…бһҠбһҖбһҮбһҪбһҡбһҲбһҡ Date бһ…бҹҒбһү бһ–бһёбһ–бҹ’бһҡбҹ„бҹҮбһҷбһҫбһ„бһҳбһ·бһ“бһ…бһ¶бҹҶбһ”бһ¶бһ…бҹӢбһ”бҹ’бһҡбһҫбһңбһ¶бһ‘бҹҖбһҸбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһ”бһ“бҹ’бһҸбһ‘бҹҒбҹ”


```r
# Load lubridate
library(lubridate)

pumpkins <- pumpkins %>% 
  # Convert the Date column to a date object
  mutate(Date = mdy(Date)) %>% 
  # Extract month from Date
  mutate(Month = month(Date)) %>% 
  # Drop Date column
  select(-Date)

# View the first few rows
pumpkins %>% 
  slice_head(n = 7)
```

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖ бһҷбһҫбһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҮбһҪбһҡбһҠбҹҒбһҖбһҗбҹ’бһҳбһёбһҳбһҪбһҷбһҲбҹ’бһҳбҹ„бҹҮ `Price` бһҠбҹӮбһӣбһҸбҹҶбһҺбһ¶бһ„бһұбҹ’бһҷбһҸбһҳбҹ’бһӣбҹғбһҮбһ¶бһҳбһ’бҹ’бһҷбһҳбһ“бҹғбһҹбҹ’бһ–бһ¶бһ“бһ–бһјбһ–бҹҒбһүбҹ” бһҘбһЎбһјбһңбһ“бҹҒбҹҮ бһҷбһҫбһ„бһ“бһ№бһ„бһӮбһҺбһ“бһ¶бһҳбһ’бҹ’бһҷбһҳбһҸбһҳбҹ’бһӣбҹғбһ–бһёбһҮбһҪбһҡбһҠбҹҒбһҖ `Low Price` бһ“бһ·бһ„ `High Price` бһҠбһҫбһҳбҹ’бһ”бһёбһ”бҹҶбһ–бҹҒбһүбһҮбһҪбһҡбһҠбҹҒбһҖбһҗбҹ’бһҳбһё Priceбҹ”  
<br>


```r
# Create a new column Price
pumpkins <- pumpkins %>% 
  mutate(Price = (`Low Price` + `High Price`)/2)

# View the first few rows of the data
pumpkins %>% 
  slice_head(n = 5)
```

бһҷбҹүбҹҒбһҹ!рҹ’Ә

"бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһҡбһ„бһ…бһ¶бҹҶ!", бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһ“бһ·бһҷбһ¶бһҷбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ”бһ¶бһ“бһҡбҹҶбһӣбһ„бһҸбһ¶бһҳбһ•бҹ’бһ‘бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ‘бһ¶бҹҶбһ„бһҳбһјбһӣбһҮбһ¶бһҳбһҪбһҷ `View(pumpkins)`, "бһҳбһ¶бһ“бһўбҹ’бһңбһёбһҳбһҪбһҷбһ…бһҳбҹ’бһӣбҹӮбһҖбһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮ!"рҹӨ”

бһ”бһҫбһўбҹ’бһ“бһҖбһҳбһҫбһӣбһ‘бҹ…бһҖбһ¶бһ“бҹӢбһҮбһҪбһҡбһҲбһҡбһҠбҹӮбһӣбһҳбһ¶бһ“бһҲбҹ’бһҳбҹ„бҹҮ `Package`, бһҳбһҫбһӣбһңбһ·бһүбһҗбһ¶вҖӢбһ‘бҹҶбһ“бһ·бһүбһ“бҹҶбһғбҹ’бһҳбһ»бҹҶбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһӣбһҖбҹӢбһҖбҹ’бһ“бһ»бһ„бһӣбһҖбҹ’бһҒбһҺбҹҲбһ•бҹ’бһҹбҹҒбһ„бҹ—бһӮбҹ’бһ“бһ¶бһ…бҹ’бһҡбһҫбһ“бҹ” бһҳбһ¶бһ“бһҒбҹ’бһӣбҹҮбһӣбһҖбҹӢбһҖбҹ’бһ“бһ»бһ„бһҳбһ¶бһҢ `1 1/9 bushel` бһ бһҫбһҷбһҒбҹ’бһӣбҹҮбһӣбһҖбҹӢбһҖбҹ’бһ“бһ»бһ„бһҳбһ¶бһҢ `1/2 bushel`, бһҒбҹ’бһӣбҹҮбһӣбһҖбҹӢбһӮбһ·бһҸбһҮбһ¶вҖӢбһ•бҹ’бһўбҹӮбһҳбһҳбҹ’бһҸбһ„бҹ—, бһҒбҹ’бһӣбҹҮбһӣбһҖбҹӢбһӮбһ·бһҸбһҮбһ¶бһӮбһёбһЎбһјбһҖбҹ’бһҡбһ¶бһҳ, бһ“бһ·бһ„бһҒбҹ’бһӣбҹҮбһҖбҹ’бһ“бһ»бһ„бһ”бҹ’бһҡбһўбһ”бҹӢбһ’бҹҶбҹ—бһҠбҹӮбһӣбһҳбһ¶бһ“бһ‘бһ‘бһ№бһ„бһҒбһ»бһҹбҹ—бһӮбҹ’бһ“бһ¶бҹ”

бһІбҹ’бһҷбһҷбһҫбһ„бһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһІбҹ’бһҷбһ”бһ¶бһ“бһ…бҹ’бһ”бһ¶бһҹбҹӢбҹ–


```r
# Verify the distinct observations in Package column
pumpkins %>% 
  distinct(Package)
```

бһҖбһ¶бһҡбҹүбһјбһҸбһ‘бҹҶбһ–бһ¶бҹҶбһ„бһ”бһ„бҹ’бһ бһ¶бһүбһҗбһ¶бһҖбһ¶бһҡбһңбһ¶бһҹбҹӢбһ‘бҹҶбһ„бһ“бҹӢбһІбҹ’бһҷбһҸбҹ’бһҡбһ№бһҳбһҸбҹ’бһҡбһјбһңбһҮбһ¶бһ”бҹ’бһҡбһ…бһ¶бҹҶбһӮбһәбһ–бһ·бһ”бһ¶бһҖбһҺбһ¶бһҹбҹӢ бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһҹбһјбһҳбһҸбһҳбҹ’бһҡбһ¶бһҳбҹ’бһҸбһ„бһҸбҹӮбһҖбһ¶бһҡбҹүбһјбһҸбһ‘бҹҶбһ–бһ¶бҹҶбһ„бһҠбҹӮбһӣбһҳбһ¶бһ“бһҒбҹ’бһҹбҹӮбһўбһҖбҹ’бһҹбһҡбһ–бһ¶бһҖбҹ’бһҷ *bushel* бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҮбһҪбһҡбһҲбһҡ `Package` бһ бһҫбһҷбһҠбһ¶бһҖбҹӢбһңбһ¶бһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„ data frame бһҗбҹ’бһҳбһёбһҲбҹ’бһҳбҹ„бҹҮ `new_pumpkins`бҹ”

#### dplyr::filter() бһ“бһ·бһ„ stringr::str_detect()

[`dplyr::filter()`](https://dplyr.tidyverse.org/reference/filter.html): бһ”бһ„бҹ’бһҖбһҫбһҸ subset бһ“бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹӮбһӣбһҳбһ¶бһ“бһҸбҹӮ **бһҮбһҪбһҡ** бһҠбҹӮбһӣбһ”бҹҶбһ–бҹҒбһүбһӣбһҖбҹ’бһҒбһҒбһҺбҹ’бһҢбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҡбһҺбһёбһ“бҹҒбҹҮ бһӮбһәбһҖбҹ’бһҡбһјбһ…бһҳбһҪбһҷбһҠбҹӮбһӣбһҳбһ¶бһ“бһҒбҹ’бһҹбҹӮбһўбһҖбҹ’бһҹбһҡ *bushel* бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҹбҹ’бһҡбһЎбһ¶бһҷ `Package` бҹ”

[stringr::str_detect()](https://stringr.tidyverse.org/reference/str_detect.html): бһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһҖбһ¶бһҡбһҸбҹ’бһҡбһјбһңбһӮбҹ’бһ“бһ¶ бһ¬бһўбһңбһҸбҹ’бһҸбһҳбһ¶бһ“бһ“бҹғбһӣбҹҶбһ“бһ¶бҹҶбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҒбҹ’бһҹбҹӮбһўбһҖбҹ’бһҹбһҡбҹ”

бһҖбһүбҹ’бһ…бһ”бҹӢ [`stringr`](https://github.com/tidyverse/stringr) бһ•бҹ’бһҸбһӣбҹӢбһҮбһјбһ“бһ“бһјбһңбһҳбһ»бһҒбһ„бһ¶бһҡвҖӢбһҠбҹӮбһӣбһҹбһ¶бһҳбһүбҹ’бһүбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҒбҹ’бһҹбҹӮбһўбһҖбҹ’бһҹбһҡбһ–бҹҒбһүбһ“бһ·бһҷбһҳбҹ”


```r
# Retain only pumpkins with "bushel"
new_pumpkins <- pumpkins %>% 
       filter(str_detect(Package, "bushel"))

# Get the dimensions of the new data
dim(new_pumpkins)

# View a few rows of the new data
new_pumpkins %>% 
  slice_head(n = 5)
```

бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҳбһҫбһӣбһғбһҫбһүбһҗбһ¶бһҷбһҫбһ„бһ”бһ¶бһ“бһҖбһ¶бһҸбҹӢбһ”бһ“бҹ’бһҗбһҷбһҠбһӣбҹӢбһ”бҹ’бһҡбһ бҹӮбһӣ бҹӨбҹЎбҹҘ бһҮбһҪбһҡбһҠбҹҒбһҸбһ¶бһҠбҹӮбһӣбһҳбһ¶бһ“бһ‘бһ“бҹ’бһҹбҹ„бһҳбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҸбһ¶бһҳбһ”бҹ’бһҡбһўбһ”бҹӢбҹ”рҹӨ©
<br>


#### dplyr::case_when()

**бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһҡбһ„бһ…бһ¶бҹҶ! бһҳбһ¶бһ“бһҡбһҝбһ„бһҳбһҪбһҷбһ‘бҹҖбһҸбһҸбҹ’бһҡбһјбһңбһ’бҹ’бһңбһҫ**

бһҸбһҫбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһҳбһҫбһӣбһғбһҫбһүбһ‘бҹҒбһҗбһ¶ бһ…бҹҶбһ“бһҪбһ“ bushel бһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡбһҸбһ¶бһҳбһҮбһҪбһҡбһҠбҹҒбһҖбһҳбһҪбһҷбҹ—? бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҸбҹӮбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһ”бҹ’бһҡбһҖбһ¶бһҹбһҸбһҳбҹ’бһӣбҹғбһұбҹ’бһҷбһҹбһҳбһҡбһҳбҹ’бһҷ бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһ бһ¶бһүбһҸбһҳбҹ’бһӣбҹғбһҖбҹ’бһ“бһ»бһ„бһҳбһҪбһҷ bushel бһҳбһ·бһ“бһҳбҹӮбһ“бһҖбҹ’бһ“бһ»бһ„ 1 1/9 бһ¬ 1/2 bushel бһ‘бҹҒбҹ” бһ–бҹҒбһӣбһңбҹҒбһӣбһ¶бһ‘бҹ…бһ’бҹ’бһңбһҫбһӮбһҺбһ·бһҸбһңбһ·бһ‘бҹ’бһҷбһ¶бһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫбһұбҹ’бһҷбһңбһ¶бһ’бһҳбҹ’бһҳбһҸбһ¶бҹ”

бһҷбһҫбһ„бһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһҳбһ»бһҒбһ„бһ¶бһҡ [`case_when()`](https://dplyr.tidyverse.org/reference/case_when.html) бһҠбһҫбһҳбҹ’бһ”бһё *бһ”бһҳбҹ’бһӣбҹӮбһ„* бһҮбһҪбһҡбһҲбһҡбһҸбһҳбҹ’бһӣбҹғ Price бһҠбҹ„бһҷбһ•бҹ’бһўбҹӮбһҖбһӣбһҫбһӣбһҖбҹ’бһҒбһҒбһҺбҹ’бһҢбһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢбһҒбҹ’бһӣбҹҮбҹ—бҹ” `case_when` бһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһұбҹ’бһҷбһўбҹ’бһ“бһҖбһ’бҹ’бһңбһҫбһңбҹүбһ·бһ…бһ‘бҹҗбһҡбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһ“бҹғ `if_else()`бҹ”


```r
# Convert the price if the Package contains fractional bushel values
new_pumpkins <- new_pumpkins %>% 
  mutate(Price = case_when(
    str_detect(Package, "1 1/9") ~ Price/(1 + 1/9),
    str_detect(Package, "1/2") ~ Price/(1/2),
    TRUE ~ Price))

# View the first few rows of the data
new_pumpkins %>% 
  slice_head(n = 30)
```

бһҘбһЎбһјбһңбһ“бҹҒбҹҮ бһҷбһҫбһ„бһўбһ¶бһ…бһңбһ·бһ—бһ¶бһӮбһҸбһҳбҹ’бһӣбҹғбһҖбҹ’бһ“бһ»бһ„бһҳбһҪбһҷбһҜбһҖбһҸбһ¶бһҸбһ¶бһҳбһҖбһ¶бһҡбһңбһ¶бһҹбҹӢбһҸбһҳбҹ’бһӣбҹғбһҳбҹүбҹӮбһҸбҹ’бһҡбһ”бҹҠбһјбһҹбҹӮбһӣбһҡбһ”бһҹбҹӢбһ–бһҪбһҖбһңбһ¶бҹ” бһҖбһ¶бһҡбһҹбһ·бһҖбҹ’бһҹбһ¶бһҡбһўбҹҶбһ–бһёбһ”бҹҠбһјбһҹбҹӮбһӣбһ“бҹғбһҖбҹ’бһҡбһјбһ…бһ”бҹҠбһәбһҸбһ‘бһ¶бҹҶбһ„бһ“бҹҒбҹҮ бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮ бһ”бһ„бҹ’бһ бһ¶бһүбһҗбһ¶ бһңбһ¶ `бһҹбҹҶбһҒбһ¶бһ“бҹӢ`бһҷбҹүбһ¶бһ„бһҒбҹ’бһӣбһ¶бҹҶбһ„бһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ‘бһ‘бһҪбһӣбһҷбһҖ `бһҖбһ¶бһҡбһҷбһӣбҹӢбһҠбһ№бһ„бһ–бһёбһ’бһҳбҹ’бһҳбһҮбһ¶бһҸбһ·бһҜбһҖбһҹбһ¶бһҡбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ`!

> вң… бһҷбҹ„бһ„бһҸбһ¶бһҳ [The Spruce Eats](https://www.thespruceeats.com/how-much-is-a-bushel-1389308) бһ‘бҹҶбһ„бһ“бҹӢбһ“бҹғбһ”бҹҠбһјбһҹбҹӮбһӣбһҳбһҪбһҷбһўбһ¶бһҹбҹ’бһҡбҹҗбһҷбһӣбһҫбһ”бҹ’бһҡбһ—бҹҒбһ‘бһ•бһӣбһ·бһҸбһ•бһӣ бһ–бҹ’бһҡбҹ„бҹҮбһңбһ¶бһҮбһ¶бһҖбһ¶бһҡбһңбһ¶бһҹбҹӢбһҸбһҳбҹ’бһӣбҹғбһ”бһҡбһ·бһҳбһ¶бһҺбҹ” "бһ”бҹҠбһјбһҹбҹӮбһӣбһҳбһҪбһҷбһ“бҹғбһ”бҹүбҹҒбһ„бһ”бҹүбҹ„бҹҮ бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһӮбһҪбһҡбһҸбҹӮбһҳбһ¶бһ“бһ‘бҹҶбһ„бһ“бҹӢбҹҘбҹҰбһ•бҹ„бһ“... бһҹбҹ’бһӣбһ№бһҖ бһ“бһ·бһ„бһҹбҹ’бһҡбһҹбҹӢбһҖбһ¶бһ“бҹӢбһҖбһ“бҹ’бһӣбҹӮбһ„бһ…бҹ’бһҡбһҫбһ“бһҮбһ¶бһҳбһҪбһҷбһ‘бҹҶбһ„бһ“бҹӢбһҸбһ·бһ…бһҮбһ¶бһ„ бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһ”бҹҠбһјбһҹбҹӮбһӣбһҳбһҪбһҷбһ“бҹғбһҹбҹ’бһ–бҹғбһҸбҹ’бһҡбһјбһңбһҳбһ¶бһ“бһ‘бҹҶбһ„бһ“бҹӢбһҸбҹ’бһҡбһ№бһҳбҹўбҹ бһ•бҹ„бһ“бһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮбҹ”" бһңбһ¶вҖӢбһ–бһ·бһ”бһ¶бһҖвҖӢбһҺбһ¶бһҹбҹӢ! бһҳбһ·бһ“бһ…бһ¶бҹҶбһ”бһ¶бһ…бҹӢбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһ”бһҳбҹ’бһӣбҹӮбһ„бһ–бһёбһ”бҹҠбһјбһҹбҹӮбһӣбһ‘бҹ…бһ•бҹ„бһ“бһ‘бҹҒ бһҒбһҺбҹҲбһҠбҹӮбһӣбһҷбһҫбһ„бһҸбҹҶбһӣбҹғбһҸбһ¶бһҳбһ”бҹҠбһјбһҹбҹӮбһӣбһ•бһ„бһҠбҹӮбһҡбҹ” бһҖбһ¶бһҡбһҹбһ·бһҖбҹ’бһҹбһ¶бһҡбһўбҹҶбһ–бһёбһ”бҹҠбһјбһҹбҹӮбһӣбһ“бҹғбһҖбҹ’бһҡбһјбһ…бһ”бҹҠбһәбһҸбһ‘бһ¶бҹҶбһ„бһ“бҹҒбҹҮ бһ”бһ„бҹ’бһ бһ¶бһүбһҷбҹүбһ¶бһ„бһҒбҹ’бһӣбһ¶бҹҶбһ„бһҗбһ¶ бһңбһ¶вҖӢбһҹбҹҶбһҒбһ¶бһ“бҹӢбһҺбһ¶бһҹбҹӢбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҷбһӣбҹӢбһҠбһ№бһ„бһ–бһёбһ’бһҳбҹ’бһҳбһҮбһ¶бһҸбһ·бһҜбһҖбһҹбһ¶бһҡбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ!
>
> вң… бһҸбһҫбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһҹбһҳбҹ’бһӮбһ¶бһӣбҹӢбһ‘бҹҒбһҗбһ¶ бһҖбҹ’бһҡбһјбһ…бһ”бҹҠбһәбһҸбһҠбҹӮбһӣбһӣбһҖбҹӢбһҸбһ¶бһҳбһҖбҹҶбһҡбһ·бһҸбһ–бһ¶бһҖбҹӢбһҖбһҺбҹ’бһҸбһ¶бһӣбһ”бҹҠбһјбһҹбҹӮбһӣ бһӮбһәбһҳбһ¶бһ“бһҸбһҳбҹ’бһӣбҹғбһҗбҹ’бһӣбҹғбһҺбһ¶бһҹбҹӢ? бһҸбһҫбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҹбҹ’бһҡбһ„бҹӢбһҷбһҖбһҳбһјбһӣбһ бҹҒбһҸбһ»бһ”бһ¶бһ“бһ‘бҹҒ? бһҮбҹҶбһ“бһҪбһҷбҹ– бһҖбҹ’бһҡбһјбһ…бһ”бҹҠбһәбһҸбһҸбһјбһ…бҹ—бһҳбһ¶бһ“бһҸбһҳбҹ’бһӣбҹғбһҗбҹ’бһӣбҹғбһҮбһ¶бһ„бһҖбҹ’бһҡбһјбһ…бһ’бҹҶ бһҠбҹ„бһҷбһҹбһ¶бһҡбһҸбҹӮбһҳбһ¶бһ“бһ…бҹҶбһ“бһҪбһ“бһҡбһ¶бһ”бҹӢбһҳбһ·бһ“бһ”бһ¶бһ“бһ…бҹ’бһҡбһҫбһ“бһҖбҹ’бһ“бһ»бһ„бһҳбһҪбһҷбһ”бҹҠбһјбһҹбҹӮбһӣ бһ”бһҫбһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бһ‘бҹ…бһ“бһ№бһ„бһҖбһ“бҹ’бһӣбҹӮбһ„бһҠбҹӮбһӣбһҳбһ·бһ“бһ”бһ¶бһ“бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҠбҹ„бһҷбһҖбҹ’бһҡбһјбһ…бһ”бһ¶бһҷбһ’бҹҶбһҳбһҪбһҷбһҠбҹӮбһӣбһ‘бһ‘бһ№бһ„бһ‘бһјбһӣбһ¶бһҷбҹ”
<br>


бһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷбһ“бҹҒбҹҮ бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ•бҹ’бһҹбһ„бһ–бҹ’бһҡбҹҒбһ„ рҹ’ҒвҖҚвҷҖпёҸ бһҸбҹ„бҹҮбһҷбһҫбһ„бһ”бҹ’бһҠбһјбһҡбһҸбҹҶбһҺбҹӮбһ„бһҖбһјбһЎбһ»бҹҶбһҒбҹӮбһ‘бҹ…бһҮбһ¶бһ‘бһёбһҸбһ¶бҹҶбһ„бһҠбҹҶбһ”бһјбһ„ бһӮбһә `бһҳбһ»бһ“` бһҖбһјбһЎбһ»бҹҶ `Package`бҹ”

`dplyr::relocate()` бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бҹ’бһҡбһҫбһҠбһҫбһҳбҹ’бһ”бһёбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡбһ‘бһёбһҸбһ¶бҹҶбһ„бһҖбһјбһЎбһ»бҹҶбҹ”


```r
# Create a new data frame new_pumpkins
new_pumpkins <- new_pumpkins %>% 
  relocate(Month, .before = Package)

new_pumpkins %>% 
  slice_head(n = 7)
```

бһҖбһ¶бһҡбһ„бһ¶бһҡбһӣбҹ’бһў!рҹ‘Ң бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһҜбһҖбһҹбһ¶бһҡбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҹбҹ’бһўбһ¶бһҸ бһ“бһ·бһ„бһҹбҹ’бһўбһ·бһҸ бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҳбҹүбһјбһҢбҹӮбһӣбһҡбҹҒбһ бҹ’бһӮбҹ’бһҡбҹҒбһҹбҹ’бһҷбһ»бһ„бһҗбҹ’бһҳбһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ”бһ¶бһ“!

> "бһҖбҹ’бһҡбһ¶бһ•бһ·бһҖбһҹбһ¶бһҳбһүбҹ’бһүбһ”бһ¶бһ“бһ“бһ¶бҹҶбһӮбҹҶбһ“бһ·бһҸбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһ…бҹ’бһҡбһҫбһ“бһ‘бҹ…бһұбҹ’бһҷбһўбҹ’бһ“бһҖбһңбһ·бһ—бһ¶бһӮбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҮбһ¶бһ„бһ§бһ”бһҖбһҡбһҺбҹҚбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбҹ”" --- John Tukey

бһ•бҹ’бһ“бҹӮбһҖбһҳбһҪбһҷбһ“бҹғбһҸбһҪбһ“бһ¶бһ‘бһёбһўбҹ’бһ“бһҖбһңбһ·бһ‘бҹ’бһҷбһ¶бһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһӮбһәбһ”бһ„бҹ’бһ бһ¶бһүбһӮбһ»бһҺбһ—бһ¶бһ–бһ“бһ·бһ„бһ’бһҳбҹ’бһҳбһҮбһ¶бһҸбһ·бһ“бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹӮбһӣбһ–бһҪбһҖбһӮбҹҒбһҹжӯЈеңЁбһ”бҹүбҹҮбһ–бһ¶бһӣбҹӢбҹ” бһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһ“бҹҒбҹҮ бһ–бһҪбһҖбһӮбҹҒбһҮбһ¶бһүбһ№бһҖбһүбһ¶бһ”бҹӢбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһ¶бһҡбһ”бһ„бҹ’бһ бһ¶бһүбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹӮбһӣбһӮбһҪбһҡбһұбҹ’бһҷбһ…бһ¶бһ”бҹӢбһўбһ¶бһҡбһҳбҹ’бһҳбһҺбҹҚ бһ¬бһ•вҖӢбһӣбһҸ бһ“бһ·бһ„бһҖбҹ’бһҡбһ¶бһ•бһ·бһҖ бһ•бҹ’бһ‘бһ¶бҹҶбһ„бһ…бһҸбҹӢ бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ•бҹ’бһҹбҹҒбһ„бҹ—бһ“бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбҹ” бһҸбһ¶бһҳбһҡбһ”бҹҖбһ”бһ“бҹҒбҹҮ бһ–бһҪбһҖбһӮбҹҒбһ…бҹӮбһҖбһ”бһ„бҹ’бһ бһ¶бһүбһ‘бҹҶбһ“бһ¶бһҖбҹӢбһ‘бҹҶбһ“бһ„ бһ“бһ·бһ„бһ…бһ“бҹ’бһӣбҹ„бҹҮбһҠбҹӮбһӣбһӣбҹҶбһ”бһ¶бһҖбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҡбһҖбһғбһҫбһүбҹ”

бһҖбһ¶бһҡбһ”бһ„бҹ’бһ бһ¶бһүбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҖбҹҸбһўбһ¶бһ…бһҮбһҪбһҷбһҖбҹҶбһҺбһҸбҹӢбһ”бһ…бҹ’бһ…бҹҒбһҖбһ‘бҹҒбһҹбһҡбҹҖбһ“бһҳбҹүбҹүбһ¶бһҹбҹҠбһёбһ“бһҠбҹӮбһӣбһҹбһҳбһҹбҹ’бһҡбһ”бһ”бҹҶбһ•бһ»бһҸбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбҹ” бһ•бҹ’бһ‘бһ¶бҹҶбһ„бһ…бһҸбҹӢбһҠбҹӮбһӣбһ бһҫбһҳбһҸбҹ’бһҡбһ¶бҹҶбһҸбһ¶бһҳбһ”бһ“бҹ’бһ‘бһ¶бһҸбҹӢбһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһ”бһ„бҹ’бһ бһ¶бһүбһҗбһ¶бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҮбһ¶бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһӣбҹ„бһҳбһӣбҹ’бһўбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ бһ¶бһҸбҹӢбһҖбһ¶бһҡбһҡбҹҖбһ“бһ”бһ“бҹ’бһ‘бһ¶бһҸбҹӢбһҸбҹ’бһҡбһ„бҹӢбҹ”

R бһ•бҹ’бһҸбһӣбҹӢбһҮбһјбһ“бһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбҹ’бһҡбһ¶бһ•бһ·бһҖ бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮ [`ggplot2`](https://ggplot2.tidyverse.org/index.html) бһӮбһәбһҮбһ¶бһҳбһҪбһҷбһҖбҹ’бһ“бһ»бһ„бһ…бҹҶбһҺбҹ„бһҳбһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһҠбҹӮбһӣбһҹбһ»бһҹбҹ’бһҡбһҹбҹӢбһ”бҹҶбһ•бһ»бһҸ бһ“бһ·бһ„бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһ”бһ¶бһ“бһ‘бһјбһӣбҹҶбһ‘бһјбһӣбһ¶бһҷбҹ” `ggplot2` бһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһұбҹ’бһҷбһўбҹ’бһ“бһҖбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбҹ’бһҡбһ¶бһ•бһ·бһҖбһҠбҹ„бһҷ **бһҖбһ¶бһҡбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһҹбһҳбһ¶бһҹбһ’бһ¶бһҸбһ»бһҜбһҖбһҡбһ¶бһҮбҹ’бһҷ**бҹ”

бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһҠбҹ„бһҷбһ•бҹ’бһ‘бһ¶бҹҶбһ„бһ…бһҸбҹӢбһҹбһ¶бһҳбһүбҹ’бһүбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҮбһҪбһҡбһҲбһҡ Price бһ“бһ·бһ„ Monthбҹ”

бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮ бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҡбһҺбһёбһ“бҹҒбҹҮ бһҷбһҫбһ„бһ“бһ№бһ„бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһҮбһ¶бһҳбһҪбһҷ [`ggplot()`](https://ggplot2.tidyverse.org/reference/ggplot.html) бһ•бҹ’бһҠбһӣбҹӢбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҮбһ¶бһӮбҹҒбһ бһ‘бҹҶбһ–бҹҗбһҡ бһ“бһ·бһ„бһҖбһ¶бһҡбһ•бҹ’бһӮбһјбһҡбһ•бҹ’бһӮбһ„бһҹбһҳбҹ’бһҡбһ„бҹӢбһўбҹҒбһҹбһ’бһёбһ‘бһёбһҖ (бһҮбһ¶бһҳбһҪбһҷ [`aes()`](https://ggplot2.tidyverse.org/reference/aes.html)) бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһ”бһүбҹ’бһ…бһјбһӣбһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢбһҳбһҪбһҷ (бһҠбһјбһ…бһҮбһ¶ [`geom_point()`](https://ggplot2.tidyverse.org/reference/geom_point.html)) бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ•бҹ’бһ‘бһ¶бҹҶбһ„бһ…бһҸбҹӢбҹ”


```r
# Set a theme for the plots
theme_set(theme_light())

# Create a scatter plot
p <- ggplot(data = new_pumpkins, aes(x = Price, y = Month))
p + geom_point()
```

бһңбһ¶бһҳбһ·бһ“бһҳбһ¶бһ“бһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбһҮбһ¶бһ–бһ·бһҹбҹҒбһҹбһ‘бҹҒ бһ–бһёбһ–бҹ’бһҡбҹ„бҹҮбһңбһ¶бһҸбҹӮбһ”бһ„бҹ’бһ бһ¶бһүбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҮбһ¶бһҮбһҪбһҡбһ•бҹ’бһҹбһ–бҹ’бһңбһ•бҹ’бһҹбһ¶бһҷбһ“бҹғбһ…бһ»бһ…бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҒбҹӮбһҠбҹӮбһӣбһ”бһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢбҹ”

### **бһҸбһҫбһҷбһҫбһ„бһ’бҹ’бһңбһҫбһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…бһҠбһҫбһҳбҹ’бһ”бһёбһІбҹ’бһҷбһңбһ¶бһҳбһ¶бһ“бһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚ?**

бһҠбһҫбһҳбҹ’бһ”бһёбһІбҹ’бһҷбһҸбһ¶бһҡбһ¶бһ„бһ”бһ„бҹ’бһ бһ¶бһүбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹӮбһӣбһҳбһ¶бһ“бһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚ бһҮбһ¶бһ‘бһјбһ‘бҹ…бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҸбҹӮбһ…бһҳбҹ’бһҡбҹҖбһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҳбһҪбһҷбҹ—бһҮбһ¶бһҖбҹ’бһҡбһ»бһҳбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбһҖбҹ’бһ“бһ»бһ„бһҖбһҡбһҺбһёбһҡбһ”бһҹбҹӢбһҷбһҫбһ„ бһҖбһ¶бһҡбһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһҸбһҳбҹ’бһӣбҹғбһҳбһ’бҹ’бһҷбһҳбһ“бҹғбһ•бҹ’бһӣбҹӮбһ…бҹҒбһҖбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҒбҹӮбһҮбһјбһ“бһўбһ¶бһ…бһ•бҹ’бһҸбһӣбҹӢбһҖбһ¶бһҡбһҷбһӣбҹӢбһҠбһ№бһ„бһ”бһ“бҹ’бһҗбҹӮбһҳбһӣбһҫбһӣбҹҶбһ“бһ¶бҹҶбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһҖбҹ’бһ“бһ»бһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҡбһ”бһҹбҹӢбһҷбһҫбһ„бҹ” бһ“бҹҒбҹҮбһ“бһ¶бҹҶбһІбҹ’бһҷбһҷбһҫбһ„бһ‘бҹ…бһҡбһҖбһҖбһ¶бһҡбһ”бһҖбһҹбҹ’бһҡбһ¶бһҷ **dplyr** бһҳбһҪбһҷбһ‘бҹҖбһҸбҹ–

#### `dplyr::group_by() %>% summarize()`

бһҖбһ¶бһҡбһ”бһјбһҖбһҹбһҡбһ»бһ”бһҸбһ¶бһҳбһҖбҹ’бһҡбһ»бһҳбһҖбҹ’бһ“бһ»бһ„ R бһўбһ¶бһ…бһӮбһҺбһ“бһ¶бһ”бһ¶бһ“бһҷбҹүбһ¶бһ„бһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ

`dplyr::group_by() %>% summarize()`

- `dplyr::group_by()` бһ”бҹ’бһҠбһјбһҡбһҜбһҖбһҸбҹ’бһҸбһ¶бһ“бҹғбһҖбһ¶бһҡбһңбһ·бһ—бһ¶бһӮбһ–бһёбһҜбһҖбһҸбҹ’бһҸбһ¶бһ“бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ‘бһ¶бҹҶбһ„бһҳбһјбһӣбһ‘бҹ…бһҮбһ¶вҖӢбһҖбҹ’бһҡбһ»бһҳвҖӢбһӣбһҖбҹ’бһҒбһҺбҹҲвҖӢбһҠбһјбһ…вҖӢбһҮбһ¶вҖӢбһҸбһ¶бһҳбһҒбҹӮбҹ”

- `dplyr::summarize()` бһ”бһ„бҹ’бһҖбһҫбһҸбһҹбҹҠбһ»бһҳбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҗбҹ’бһҳбһёбһҠбҹӮбһӣбһҳбһ¶бһ“бһҳбһҪбһҷбһҮбһҪбһҡбһҲбһҡбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бҹүбһ¶бһҡбҹүбһ¶бһҳбҹүбҹӮбһҸбҹ’бһҡбһҖбҹ’бһҡбһ»бһҳбһ“бһёбһҳбһҪбһҷ бһ“бһ·бһ„бһҳбһҪбһҷбһҮбһҪбһҡбһҲбһҡбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҹбҹ’бһҗбһ·бһҸбһ·бһҡбһҪбһҳбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбҹ”

бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһҷбһҫбһ„бһўбһ¶бһ…бһ”бҹ’бһҡбһҫ `dplyr::group_by() %>% summarize()` бһҠбһҫбһҳбҹ’бһ”бһёбһ…бһҳбҹ’бһҡбҹҖбһ„бһ•бҹ’бһӣбҹӮбһ…бҹҒбһҖбһ‘бҹ…бһҮбһ¶бһҖбҹ’бһҡбһ»бһҳбһҮбһ¶бһҸбһ¶бһҳбһҮбһҪбһҡбһҲбһҡ **Month** бһ бһҫбһҷбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһҹбҹ’бһңбҹӮбһ„бһҡбһҖ **бһҸбһҳбҹ’бһӣбҹғбһҳбһ’бҹ’бһҷбһҳ** бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҡбһ¶бһ”бҹӢбһҒбҹӮбһҡбһ”бһҹбҹӢбһҳбһҪбһҷбҹ—бҹ”


```r
# Find the average price of pumpkins per month
new_pumpkins %>%
  group_by(Month) %>% 
  summarise(mean_price = mean(Price))
```

бһҹбһ„бҹ’бһҒбҹҒбһ”!вңЁ

бһӣбһҖбҹ’бһҒбһҺбҹҲбһҮбһ¶бһҖбҹ’бһҡбһ»бһҳбһҠбһјбһ…бһҮбһ¶бһҒбҹӮбһҖбһ¶бһ“бҹӢбһҸбҹӮбһӣбҹ’бһўбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ‘бҹ’бһҡбһ‘бҹ’бһҡбһ„бҹӢбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҖбҹ’бһҡбһ¶бһ”бһ”бҹүбһ¶бһҡбҹүбҹӮрҹ“Ҡбҹ” бһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢбһҠбҹӮбһӣбһ‘бһ‘бһҪбһӣбһҒбһ»бһҹбһҸбҹ’бһҡбһјбһңбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбҹ’бһҡбһ¶бһ”бһ”бҹүбһ¶бһҡбҹүбҹӮбһӮбһә `geom_bar()` бһ“бһ·бһ„ `geom_col()`бҹ” бһҹбһјбһҳбһ–бһ·бһ“бһ·бһҸбҹ’бһҷ `?geom_bar` бһҠбһҫбһҳбҹ’бһ”бһёбһҹбҹ’бһңбҹӮбһ„бһҷбһӣбҹӢбһ”бһ“бҹ’бһҗбҹӮбһҳбҹ”

бһҳбһҖбһ’бҹ’бһңбһҫбһҳбһҪбһҷ!


```r
# Find the average price of pumpkins per month then plot a bar chart
new_pumpkins %>%
  group_by(Month) %>% 
  summarise(mean_price = mean(Price)) %>% 
  ggplot(aes(x = Month, y = mean_price)) +
  geom_col(fill = "midnightblue", alpha = 0.7) +
  ylab("Pumpkin Price")
```
