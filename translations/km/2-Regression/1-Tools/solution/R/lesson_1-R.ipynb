## бһҖбһ¶бһҡбһҺбҹӮбһ“бһ¶бҹҶбһўбҹҶбһ–бһёбһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷ - бһҳбҹҒбһҡбҹҖбһ“ бҹЎ

#### бһҷбһҖбһңбһ¶бһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһ‘бһ·бһҹбһҠбҹ…

вң… бһҳбһ¶бһ“бһ”бҹ’бһҡбһ—бҹҒбһ‘бһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҡбҹ’бһҸбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“ бһ бһҫбһҷбһҖбһ¶бһҡбһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһўбһ¶бһҹбҹ’бһҡбҹҗбһҷбһӣбһҫбһ…бһҳбҹ’бһӣбһҫбһҷбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһҖбҹҶбһ–бһ»бһ„бһҹбҹ’бһңбҹӮбһ„бһҡбһҖбҹ” бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһ…бһ„бҹӢбһ‘бһ¶бһҷбһ‘бһ»бһҖбһҖбһҳбҹ’бһ–бһҹбҹӢбһ”бҹ’бһҡбһ бҹӮбһӣбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҳбһ“бһ»бһҹбҹ’бһҹбһҳбҹ’бһ“бһ¶бһҖбҹӢбһҳбһ¶бһ“бһўбһ¶бһҷбһ»бһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһ”бҹ’бһҡбһҫ `linear regression` бһ–бһёбһ–бҹ’бһҡбҹ„бҹҮбһўбҹ’бһ“бһҖбһҖбҹҶбһ–бһ»бһ„бһҹбҹ’бһңбҹӮбһ„бһҡбһҖ **бһҸбһҳбҹ’бһӣбҹғбһӣбҹҒбһҒ**бҹ” бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһ…бҹҶбһҺбһ¶бһ”бҹӢбһўбһ¶бһҡбһҳбҹ’бһҳбһҺбҹҚбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҡбһҖбһҳбһҫбһӣбһҗбһ¶бһҸбһҫбһ”бҹ’бһҡбһ—бҹҒбһ‘бһҳбҹ’бһ бһјбһ”бһҳбһҪбһҷбһӮбһҪбһҡбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһӮбһ·бһҸбһҗбһ¶бһҮбһ¶ бһңбҹҒбһ бҹ’бһӮбһ¶бһ“бҹӢ бһ¬бһўбһҸбҹӢ бһўбҹ’бһ“бһҖбһҖбҹҶбһ–бһ»бһ„бһҹбҹ’бһңбҹӮбһ„бһҡбһҖ **бһҖбһ¶бһҡбһ”бҹӮбһ„бһ…бҹӮбһҖбһ”бҹ’бһҡбһ—бҹҒбһ‘** бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһ”бҹ’бһҡбһҫ `logistic regression`бҹ” бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“бһ”бһ“бҹ’бһҗбҹӮбһҳбһўбҹҶбһ–бһё logistic regression бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбҹ” бһҹбһјбһҳбһӮбһ·бһҸбһ”бһ“бҹ’бһҸбһ·бһ…бһўбҹҶбһ–бһёбһҹбҹҶбһҺбһҪбһҡбһҒбҹ’бһӣбҹҮбҹ—бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҹбһҪбһҡбһ–бһёбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ бһ бһҫбһҷбһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҡбҹ’бһҸбһҺбһ¶бһҳбһҪбһҷбһ“бҹғбһңбһ·бһ’бһёа№Җаё«аёҘа№ҲаёІаёҷаёөа№үбһҠбҹӮбһӣбһўбһ¶бһ…бһҹбһҳбһҡбһҳбҹ’бһҷбһҮбһ¶бһ„бһӮбҹҒбҹ”

бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ•бҹ’бһ“бҹӮбһҖбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷ [бһҢбһ¶бһҸбһ¶бһҹбҹҒбһҸбһҸбһјбһ…бһҳбһҪбһҷбһўбҹҶбһ–бһёбһҮбҹҶбһ„бһәбһ‘бһ№бһҖбһ“бҹ„бһҳбһ•бҹ’бһўбҹӮбһҳ](https://www4.stat.ncsu.edu/~boos/var.select/diabetes.html)бҹ” бһҹбһјбһҳбһ“бһ№бһҖбһҹбҹ’бһҡбһҳбҹғбһҗбһ¶ бһўбҹ’бһ“бһҖбһ…бһ„бҹӢбһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһҖбһ¶бһҡбһ–бҹ’бһҷбһ¶бһ”бһ¶бһӣбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһўбҹ’бһ“бһҖбһҮбҹҶбһ„бһәбһ‘бһ№бһҖбһ“бҹ„бһҳбһ•бҹ’бһўбҹӮбһҳбҹ” бһҳбҹүбһ¶бһҹбҹҠбһёбһ“бһҡбҹҖбһ“бһўбһ¶бһ…бһҮбһҪбһҷбһўбҹ’бһ“бһҖбһҖбҹҶбһҺбһҸбҹӢбһҗбһ¶ бһўбҹ’бһ“бһҖбһҮбҹҶбһ„бһәбһҺбһ¶бһҠбҹӮбһӣбһ“бһ№бһ„бһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһӣбҹ’бһўбһҮбһ¶бһ„бһ…бҹҶбһ–бҹ„бҹҮбһҖбһ¶бһҡбһ–бҹ’бһҷбһ¶бһ”бһ¶бһӣ бһҠбҹ„бһҷбһ•бҹ’бһўбҹӮбһҖбһӣбһҫбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһҡбһҪбһҳбһ“бҹғбһўбһҗбҹҒбһҡбһ–бһёбһҡбһ”бҹҖбһ”бһ•бҹ’бһҹбҹҒбһ„бҹ—бҹ” бһҸбҹӮбһҳбҹүбһјбһҠбҹӮбһӣбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҳбһ¶бһ“бһҸбҹҶбһҡбһјбһңбһҖбһ¶бһҡбһҹбһ¶бһҳбһүбҹ’бһүбһҳбһҪбһҷ бһҖбһ¶бһӣбһҺбһ¶бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бһ„бҹ’бһ бһ¶бһүбһҮбһ¶бһҡбһјбһ”бһ—бһ¶бһ– бһўбһ¶бһ…бһ”бһ„бҹ’бһ бһ¶бһүбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһўбҹҶбһ–бһёбһўбһҗбҹҒбһҡбһ–бһёбһҡбһҠбҹӮбһӣбһўбһ¶бһ…бһҮбһҪбһҷбһўбҹ’бһ“бһҖбһҡбҹҖбһ”бһ…бҹҶбһ”бһ‘бһ–бһ·бһҹбҹ„бһ’бһ“бҹҚбһӮбҹ’бһӣбһёбһ“бһ·бһҖбһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһ‘бҹ’бһҡбһ№бһҹбҹ’бһҸбһёбҹ”

## 1. бһҖбһ¶бһҡбһ•бҹ’бһ‘бһ»бһҖбһ§бһ”бһҖбһҡбһҺбҹҚбһҡбһ”бһҹбҹӢбһҷбһҫбһ„бһЎбһҫбһ„

бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ—бһ¶бһҡбһҖбһ·бһ…бҹ’бһ…бһ“бҹҒбҹҮ бһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

-   `tidyverse`: [tidyverse](https://www.tidyverse.org/) бһӮбһәбһҮбһ¶ [бһҖбһүбҹ’бһ…бһ”бҹӢбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ R](https://www.tidyverse.org/packages) бһҠбҹӮбһӣбһҡбһ…бһ“бһ¶бһЎбһҫбһ„бһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫбһұбҹ’бһҷбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһңбһ·бһ‘бҹ’бһҷбһ¶бһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҳбһ¶бһ“бһ—бһ¶бһ–бһҡбһ бҹҗбһҹ бһ”бҹ’бһҡбһҹбһҫбһҡбһЎбһҫбһ„ бһ“бһ·бһ„бһӮбһҪбһҡбһұбҹ’бһҷбһҹбһ”бҹ’бһ”бһ¶бһҷбһ…бһ·бһҸбҹ’бһҸ!

-   `tidymodels`: бһҖбҹ’бһҡбһ”бһҒбһҺбҹ’бһҢ [tidymodels](https://www.tidymodels.org/) бһӮбһәбһҮбһ¶ [бһҖбһүбҹ’бһ…бһ”бҹӢбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ](https://www.tidymodels.org/packages/) бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһӮбҹҶбһҡбһј бһ“бһ·бһ„бһҖбһ¶бһҡбһҹбһ·бһҖбҹ’бһҹбһ¶бһҳбҹүбһ¶бһҹбҹҠбһёбһ“бҹ”

бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҠбҹҶбһЎбһҫбһ„бһңбһ¶бһҠбһјбһ…бһҮбһ¶бҹ–

`install.packages(c("tidyverse", "tidymodels"))`

бһҹбҹ’бһӮбҹ’бһҡбһёбһ”бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбһ“бҹҒбҹҮбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҗбһ¶ бһҸбһҫбһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ‘бһ¶бҹҶбһ„бһ“бҹҒбҹҮбһҠбһҫбһҳбҹ’бһ”бһёбһ”бһүбҹ’бһ…бһ”бҹӢбһҳбһјбһҢбһ»бһӣбһ“бҹҒбҹҮбһҡбһҪбһ…бһҗбҹӮбһҳбһ‘бҹҖбһҸ бһ бһҫбһҷбһҸбҹҶбһЎбһҫбһ„бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһўбҹ’бһ“бһҖ бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһҳбһ¶бһ“бһҒбһ¶бһҸбһҒбһ¶бһҸбҹ”


```r
suppressWarnings(if(!require("pacman")) install.packages("pacman"))
pacman::p_load(tidyverse, tidymodels)
```

бһҘбһЎбһјбһңбһ“бҹҒбҹҮ бһҷбһҫбһ„бһҹбһјбһҳбһ•бҹ’бһ‘бһ»бһҖбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһўбһҹбҹ’бһ…бһ¶бһҡбҹ’бһҷбһ‘бһ¶бҹҶбһ„бһ“бҹҒбҹҮ бһ бһҫбһҷбһ’бҹ’бһңбһҫбһІбҹ’бһҷбһңбһ¶бһ”бҹ’бһҡбһҫбһ”бһ¶бһ“бһҖбҹ’бһ“бһ»бһ„бһҹбһҳбҹҗбһҷ R бһ”бһ…бҹ’бһ…бһ»бһ”бҹ’бһ”бһ“бҹ’бһ“бһҡбһ”бһҹбҹӢбһҷбһҫбһ„бҹ” (бһ“бҹҒбҹҮбһӮбҹ’бһҡбһ¶бһ“бҹӢбһҸбҹӮбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ”бһ„бҹ’бһ бһ¶бһүбһҸбҹӮбһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮ `pacman::p_load()` бһ”бһ¶бһ“бһ’бҹ’бһңбһҫбһҡбһҪбһ…бһ бһҫбһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһўбҹ’бһ“бһҖ)

> рҹҺ“ бһ…бһ„бһ…бһ¶бҹҶбһҗбһ¶ бһ“бҹҒбҹҮбһӮбһәбһҮбһ¶бһҖбһ¶бһҡбһҡбҹҖбһ“бһҠбҹ„бһҷбһҳбһ¶бһ“бһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷ бһ бһҫбһҷбһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһӮбҹ„бһӣбһҠбҹ…бһҲбҹ’бһҳбҹ„бҹҮ 'y'бҹ”

бһҳбһ»бһ“бһ–бҹҒбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҮбһ¶бһҳбһҪбһҷ R бһ”бһ¶бһ“ бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһ“бһ¶бҹҶбһ…бһјбһӣбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһўбһ„бҹ’бһӮбһ…бҹҗбһ“бҹ’бһ‘ R бһ¬ бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһ¶бһҡбһҸбһ—бҹ’бһҮбһ¶бһ”бҹӢбһ‘бҹ…бһҖбһ¶бһ“бҹӢбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹӮбһӣ R бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһҠбһҫбһҳбҹ’бһ”бһёбһ…бһјбһӣбһ”бҹ’бһҡбһҫбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ–бһёбһ…бһҳбҹ’бһ„бһ¶бһҷбҹ”

> бһҖбһүбҹ’бһ…бһ”бҹӢ [readr](https://readr.tidyverse.org/), бһҠбҹӮбһӣбһҮбһ¶бһ•бҹ’бһ“бҹӮбһҖбһҳбһҪбһҷбһ“бҹғ Tidyverse, бһ•бҹ’бһҠбһӣбҹӢбһ“бһјбһңбһңбһ·бһ’бһёбһҡбһ бҹҗбһҹ бһ“бһ·бһ„бһ• friendliness бһ”бһ¶бһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһўбһ¶бһ“бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҡбһ¶бһ„бһӮбһјбһ‘бҹҗбһҡбһ…бһјбһӣбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„ Rбҹ”

бһҘбһӣбһјбһңбһ“бҹҒбҹҮ бһ“бһ¶бҹҶбһ…бһјбһӣбһҹбҹҶбһҺбһ»бҹҶбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҮбҹҶбһ„бһәбһҶбҹ’бһӣбһ„бһҲ СҒР°С…Р°СҖ бһҠбҹӮбһӣбһ•бҹ’бһҠбһӣбҹӢбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„ URL бһ•бҹ’бһҠбһҫбһҳбһ“бҹҒбҹҮбҹ– <https://www4.stat.ncsu.edu/~boos/var.select/diabetes.html>

бһ”бһ“бҹ’бһҗбҹӮбһҳбһ–бһёбһ“бҹҒбҹҮ бһҷбһҫбһ„бһ“бһ№бһ„вҖӢбһ’бҹ’бһңбһҫбһҸбҹҒбһҹбҹ’бһҸбһҹбһ»бһңбһҸбҹ’бһҗбһ·бһ—бһ¶бһ–бһҮбһ¶бһҳбһҪбһҷбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҡбһ”бһҹбҹӢбһҷбһҫбһ„бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ `glimpse()` бһ бһҫбһҷбһ”бһ„бҹ’бһ бһ¶бһүбһҮбһҪбһҡбһҠбҹҒбһҖбһҠбҹҶбһ”бһјбһ„ бҹҘ бһҮбһҪбһҡбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ `slice()`бҹ”

бһҳбһ»бһ“бһ–бҹҒбһӣбһ‘бҹ…бһҖбһ¶бһ“бҹӢбһҮбҹҶбһ бһ¶бһ“бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢ бһҹбһјбһҳбһҺбҹӮбһ“бһ¶бҹҶбһўбҹ’бһңбһёбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҮбһҪбһ”бһүбһ№бһҖбһүбһ¶бһ”бҹӢбһҖбҹ’бһ“бһ»бһ„бһҖбһјбһҠ R рҹҘҒрҹҘҒ: бһўбһ»бһ”бһҡбҹүбһ¶бһҸбҹҗбһҡбһ–бһёбһ” `%>%`

бһўбһ»бһ”бһҡбҹүбһ¶бһҸбҹҗбһҡбһ–бһёбһ” (`%>%`) бһ”бҹҶбһ–бҹҒбһүбһҹбһҖбһҳбҹ’бһҳбһ—бһ¶бһ–бһҮбһ¶бһҠбҹҶбһҺбһ¶бһҖбҹӢбһҖбһ¶бһӣбһҸбһ¶бһҳбһӣбҹҶбһҠбһ¶бһ”бҹӢ бһҠбҹ„бһҷбһ•бҹ’бһҠбһӣбҹӢбһңбһҸбҹ’бһҗбһ»бһ‘бҹ…бһҳбһ»бһҒбһҖбҹ’бһ“бһ»бһ„бһҳбһ»бһҒбһ„бһ¶бһҡбһ¬бһңбҹҒбһҷбҹ’бһҷбһ¶бһҖбһҡбһҺбҹҚбһ бҹ…бҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһӮбһ·бһҸбһўбһ»бһ”бһҡбҹүбһ¶бһҸбҹҗбһҡбһ–бһёбһ”бһҗбһ¶ вҖңбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖвҖқ бһҖбҹ’бһ“бһ»бһ„бһҖбһјбһҠбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбҹ”


```r
# Import the data set
diabetes <- read_table2(file = "https://www4.stat.ncsu.edu/~boos/var.select/diabetes.rwrite1.txt")


# Get a glimpse and dimensions of the data
glimpse(diabetes)


# Select the first 5 rows of the data
diabetes %>% 
  slice(1:5)
```

`glimpse()` бһ”бһ„бҹ’бһ бһ¶бһүбһІбҹ’бһҷбһҷбһҫбһ„бһғбһҫбһүбһҗбһ¶вҖӢбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ“бҹҒбҹҮбһҳбһ¶бһ“бһҮбһҪбһҡбһҠбҹҒбһҖ бҹӨбҹӨбҹў бһ“бһ·бһ„бһҮбһҪбһҡбһҲбһҡ бҹЎбҹЎ бһҠбҹӮбһӣбһҮбһҪбһҡбһҲбһҡбһ‘бһ¶бҹҶбһ„бһўбһҹбҹӢбһҳбһ¶бһ“бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҮбһ¶ `double`

> glimpse() бһ“бһ·бһ„ slice() бһӮбһәбһҮбһ¶бһўбһ“бһ»бһӮбһҳбһ“бҹҚбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„ [`dplyr`](https://dplyr.tidyverse.org/)бҹ” Dplyr бһҮбһ¶бһ•бҹ’бһ“бҹӮбһҖбһҳбһҪбһҷбһ“бҹғ Tidyverse бһӮбһәбһҮбһ¶бһңбһ·бһ”бҹ’бһ”бһҠбһ·бһҹбһ¶бһҡбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ бһҠбҹӮбһӣбһ•бҹ’бһҸбһӣбҹӢбһ“бһјбһңбһҹбҹҶбһҺбһ»бҹҶбһӮбҹҶбһҡбһјбһ–бһ¶бһҖбҹ’бһҷбһҖбһ·бһҡбһ·бһҷбһ¶бһҠбҹӮбһӣбһҳбһ¶бһ“бһ—бһ¶бһ–бһҹбҹ’бһҡбһ”бһӮбҹ’бһ“бһ¶ бһҠбҹӮбһӣбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһ”бһүбҹ’бһ бһ¶бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ‘бһјбһ‘бҹ…бһ”бҹҶбһ•бһ»бһҸ

бһҘбһЎбһјбһңбһ–бҹҒбһӣбһҠбҹӮбһӣбһҷбһҫбһ„бһҳбһ¶бһ“бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ бһҫбһҷ бһҳбһҖбһҖбһ¶бһ“бҹӢбһҖбһ¶бһҡбһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҳбһҪбһҷбһӣбһҖбҹ’бһҒбһҺбҹҲ (`bmi`) бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһўбһ“бһ»бһңбһҸбҹ’бһҸбһҖбһ¶бһҡбһ бһ¶бһҸбҹӢбһ“бҹҒбҹҮбҹ” бһ“бҹҒбҹҮбһҖбҹҸбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһІбҹ’бһҷбһҷбһҫбһ„бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҮбһҪбһҡбһҲбһҡбһҠбҹӮбһӣбһ…бһ„бҹӢбһ”бһ¶бһ“бҹ” бһҸбһҫбһҷбһҫбһ„бһ’бҹ’бһңбһҫбһ”бҹӮбһ”бһҺбһ¶?

[`dplyr::select()`](https://dplyr.tidyverse.org/reference/select.html) бһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһІбҹ’бһҷбһҷбһҫбһ„ *бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹ* (бһ бһҫбһҷбһүбһ№бһҖбһүбһ¶бһ”бҹӢбһўбһ¶бһ…бһ”бҹ’бһҸбһјбһҡбһҲбҹ’бһҳбҹ„бҹҮ) бһҮбһҪбһҡбһҲбһҡбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҹбҹҶбһҺбһ»бҹҶбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҳбһҪбһҷбҹ”


```r
# Select predictor feature `bmi` and outcome `y`
diabetes_select <- diabetes %>% 
  select(c(bmi, y))

# Print the first 5 rows
diabetes_select %>% 
  slice(1:10)
```

## 3. бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣ бһ“бһ·бһ„бһҹбһ¶бһҖбһӣбҹ’бһ”бһ„

бһңбһ¶бһҮбһ¶бһңбһ·бһ’бһ¶бһ“бһҖбһ¶бһҡбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһўбһ“бһ»бһңбһҸбҹ’бһҸбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҡбҹҖбһ“бһҸбһ¶бһҳбһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷ бһҠбһҫбһҳбҹ’бһ”бһё *бһ”бҹҶбһ”бҹӮбһҖ* бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ…бҹҒбһүбһҮбһ¶бһҹбҹҶбһҺбһ»бҹҶбһ“бҹғбһ–бһёбһҡбһ•бҹ’бһ“бҹӮбһҖ; бһҹбҹҶбһҺбһ»бҹҶбһ“бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ (бһҠбҹӮбһӣбһҮбһ¶бһ‘бһјбһ‘бҹ…бһ’бҹҶбһҮбһ¶бһ„) бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһ„бҹ’бһҡбҹҖбһ“бһҳбҹүбһјбһҠбҹӮбһӣ бһ бһҫбһҷбһҹбҹҶбһҺбһ»бҹҶбһ“бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҸбһјбһ… "hold-back" бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҳбһҫбһӣбһҗбһ¶ бһҳбҹүбһјбһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бҹҶбһ–бҹҒбһүбһҖбһ¶бһҡбһ„бһ¶бһҡбһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…бҹ”

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһ–бҹҒбһӣбһҷбһҫбһ„бһҳбһ¶бһ“бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҡбһҪбһ…бһҡбһ¶бһӣбҹӢ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҳбһҫбһӣбһҗбһ¶бһҸбһҫбһҳбҹүбһ¶бһҹбҹҠбһёбһ“бһўбһ¶бһ…бһҮбһҪбһҷбһҖбҹҶбһҺбһҸбҹӢбһҖбһ¶бһҡбһ”бҹҶбһ”бҹӮбһҖбһҸбҹ’бһҡбһ№бһҳбһҸбҹ’бһҡбһјбһңбһҡбһңбһ¶бһ„бһӣбҹҒбһҒбһҖбҹ’бһ“бһ»бһ„бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ“бҹҒбҹҮбһ”бһ¶бһ“бһ‘бҹҒбҹ” бһҷбһҫбһ„бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһҖбһүбҹ’бһ…бһ”бҹӢ [rsample](https://tidymodels.github.io/rsample/) бһҠбҹӮбһӣбһҮбһ¶бһҳбһҪбһҷбһ“бһ№бһ„бһҹбҹҠбһ»бһҳбһҡбҹүбҹӮбһ’бһҳбһјбһӣ Tidymodels бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбһ·бһ”бһҡбһҳбһ¶бһҳбһҪбһҷбһҠбҹӮбһӣбһҳбһ¶бһ“бһ–бҹ’бһҡбһ№бһҸбҹ’бһҸбһ·бһ”бһҸбҹ’бһҡбһҹбҹ’бһҸбһёбһ–бһё *бһңбһ·бһ’бһё* бһ”бҹҶбһ”бҹӮбһҖбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ бһ бһҫбһҷбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһ”бҹ’бһҡбһҫбһҳбһ»бһҒбһ„бһ¶бһҡ rsample бһ…бҹ’бһҡбһҫбһ“бһ‘бҹҖбһҸбһ–бһёбһҡбһҳбһ»бһҒбһ„бһ¶бһҡбһҠбһҫбһҳбҹ’бһ”бһёбһ‘бһ¶бһүбһҷбһҖбһҹбҹҶбһҺбһ»бҹҶбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣ бһ“бһ·бһ„бһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбҹ–


```r
set.seed(2056)
# Split 67% of the data for training and the rest for tesing
diabetes_split <- diabetes_select %>% 
  initial_split(prop = 0.67)

# Extract the resulting train and test sets
diabetes_train <- training(diabetes_split)
diabetes_test <- testing(diabetes_split)

# Print the first 3 rows of the training set
diabetes_train %>% 
  slice(1:10)
```

## 4. бһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбһӮбҹҶбһҡбһјбһҹбһҳбһёбһҖбһ¶бһҡбһҹбҹ’бһҡбһҳбҹ„бһӣбһҮбһ¶бһҳбһҪбһҷ Tidymodels

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҷбһҫбһ„бһҳбһ¶бһ“бһҹбҹ’бһҡбҹҒбһ…бһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбһӮбҹҶбһҡбһјбһҡбһ”бһҹбҹӢбһҷбһҫбһ„!

бһҖбҹ’бһ“бһ»бһ„ Tidymodels бһўбҹ’бһ“бһҖбһҖбҹҶбһҺбһҸбҹӢбһӮбҹҶбһҡбһјбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ `parsnip()` бһҠбҹ„бһҷбһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһ”бһёбһҖбһ“бҹ’бһӣбҹӮбһ„бҹ–

-   бһ”бҹ’бһҡбһ—бҹҒбһ‘бһӮбҹҶбһҡбһј **type** бһ•бҹ’бһҸбҹӮбһҖбһӮбҹҶбһҡбһјбҹ— бһҠбһјбһ…бһҮбһ¶бһҹбһҳбһёбһҖбһ¶бһҡбһҹбҹ’бһҡбһҳбҹ„бһӣбһ”бһ“бҹ’бһ‘бһ¶бһҸбҹӢ, бһҹбһҳбһёбһҖбһ¶бһҡбһҹбҹ’бһҡбһҳбҹ„бһӣбһЎбһјбһҮбһёбһҹбҹ’бһ‘бһ·бһ…, бһӮбҹҶбһҡбһјбһҠбһҫбһҳбһҹбһҳбҹ’бһҡбҹҒбһ…бһ…бһ·бһҸбҹ’бһҸ бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бҹ—бһ‘бҹҖбһҸбҹ”

-   бһҡбһ”бҹҖбһ”бһ’бҹ’бһңбһҫбһҠбҹҶбһҺбһҫбһҡ **mode** бһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһҮбһҳбҹ’бһҡбһҫбһҹбһ‘бһјбһ‘бҹ…бһҠбһјбһ…бһҮбһ¶ бһҹбһҳбһёбһҖбһ¶бһҡбһҹбҹ’бһҡбһҳбҹ„бһӣ бһ“бһ·бһ„ бһҖбһ¶бһҡбһ”бҹӮбһ„бһ…бҹӮбһҖбһ…бҹҶбһҺбһ¶бһҸбҹӢбһҗбҹ’бһ“бһ¶бһҖбҹӢ; бһ”бҹ’бһҡбһ—бҹҒбһ‘бһӮбҹҶбһҡбһјбһҳбһҪбһҷбһ…бҹҶбһ“бһҪбһ“бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһҸбҹӮбһҳбҹ’бһҠбһ„бһ¬бһ‘бһ¶бҹҶбһ„бһ–бһёбһҡ бһҒбһҺбҹҲбһҠбҹӮбһӣбһҳбһҪбһҷбһ…бҹҶбһ“бһҪбһ“бһҳбһ¶бһ“бһҸбҹӮбһҡбһ”бҹҖбһ”бһ’бҹ’бһңбһҫбһҠбҹҶбһҺбһҫбһҡбһҸбҹӮбһҳбһҪбһҷбҹ”

-   бһҳбҹҒбһҖбһ¶бһ“бһ·бһ…бһӮбҹҶбһҡбһј **engine** бһӮбһәбһҮбһ¶бһ§бһ”бһҖбһҡбһҺбҹҚбһӮбһҺбһ“бһ¶бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҸбһҳбҹ’бһҡбҹҖбһ”бһӮбҹҶбһҡбһјбҹ” бһҮбһ¶бһүбһ№бһҖбһүбһ¶бһ”бҹӢбһӮбһәбһҮбһ¶бһҖбһүбҹ’бһ…бһ”бҹӢ R бһҳбһҪбһҷбһ…бҹҶбһ“бһҪбһ“ бһҠбһјбһ…бһҮбһ¶ **`"lm"`** бһ¬ **`"ranger"`**

бһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһ“бҹҒбҹҮбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҖбһҸбҹӢбһҸбҹ’бһҡбһ¶вҖӢбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһӮбҹҶбһҡбһј бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһҳбһҖбһ”бһ„бҹ’бһҖбһҫбһҸбһҳбһҪбһҷбһ”бһ¶бһ“бһ бһҫбһҷ!


```r
# Build a linear model specification
lm_spec <- 
  # Type
  linear_reg() %>% 
  # Engine
  set_engine("lm") %>% 
  # Mode
  set_mode("regression")


# Print the model specification
lm_spec
```

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһҳбҹүбһјбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“ *бһҖбҹҶбһҺбһҸбҹӢ* бһҡбһҪбһ…бһ бһҫбһҷ бһҳбҹүбһјбһҠбҹӮбһӣбһўбһ¶бһ…бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“ `бһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһ”бҹ’бһҡбһҳбһ¶бһҺ` бһ¬ `бһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣ` бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҳбһ»бһҒбһ„бһ¶бһҡ [`fit()`](https://parsnip.tidymodels.org/reference/fit.html) бһҮбһ¶бһ‘бһјбһ‘бҹ…бһ”бҹ’бһҡбһҫбһҡбһјбһ”бһҳбһ“бҹ’бһҸбһҳбһҪбһҷбһ“бһ·бһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҳбһҪбһҷбһ…бҹҶбһ“бһҪбһ“бҹ”

`y ~ .` бһҳбһ¶бһ“бһ“бҹҗбһҷбһҗбһ¶бһҷбһҫбһ„бһ“бһ№бһ„бһҸбҹ’бһҡбһјбһңбһ”бһҺбҹ’бһҠбһ»бҹҮ `y` бһҮбһ¶бһ”бһҡбһ·бһҳбһ¶бһҺбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһ–бҹ’бһҷбһ¶бһҖбһҡбһҺбҹҚ/бһӮбҹ„бһӣбһҠбҹ… бһҠбҹӮбһӣбһ–бһҺбҹҢбһ“бһ¶бһҠбҹ„бһҷбһўбҹ’бһ“бһҖбһҠбһ¶бҹҶбһ“бҹҒбҹҮбһҡбһ¶бһӣбҹӢбһӮбҹ’бһ“бһ¶ ie, `.` (бһҖбҹ’бһ“бһ»бһ„бһҖбһҡбһҺбһёбһ“бҹҒбҹҮ бһҷбһҫбһ„бһҳбһ¶бһ“бһўбҹ’бһ“бһҖбһҠбһ¶бҹҶбһҸбҹӮбһҳбһҪбһҷбһӮбһҸбҹӢбһӮбһә `bmi` )


```r
# Build a linear model specification
lm_spec <- linear_reg() %>% 
  set_engine("lm") %>%
  set_mode("regression")


# Train a linear regression model
lm_mod <- lm_spec %>% 
  fit(y ~ ., data = diabetes_train)

# Print the model
lm_mod
```

бһ–бһёбһӣбһ‘бҹ’бһ’бһ•бһӣбһ“бҹғбһҳбҹүбһјбһҢбҹӮбһӣ бһҷбһҫбһ„бһўбһ¶бһ…бһҳбһҫбһӣбһғбһҫбһүбһўбһ“бһ»бһӮбһҳбһ“бҹҚбһҠбҹӮбһӣбһ”бһ¶бһ“бһҡбҹҖбһ“бһғбһҫбһүбһ“бҹ…бһ–бҹҒбһӣбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбҹ” бһңбһ¶бһҸбҹҶбһҺбһ№бһ„бһўбһ“бһ»бһӮбһҳбһ“бҹҚбһ“бҹғбһ”бһ“бҹ’бһ‘бһ¶бһҸбҹӢбһҠбҹӮбһӣбһҳбһ¶бһ“бһҖбһ¶бһҡбһҹбһҳбҹ’бһҡбһҪбһӣбһӣбҹ’бһўбһ”бҹҶбһ•бһ»бһҸбһҠбҹӮбһӣбһ•бҹ’бһҸбһӣбҹӢбһұбҹ’бһҷбһҷбһҫбһ„бһ“бһјбһңбһҖбҹҶбһ бһ»бһҹбһҹбһҡбһ»бһ”бһ‘бһ¶бһ”бһҮбһ¶бһ„бһӮбҹҒбһ‘бһ¶бҹҶбһ„бһ–бһёбһҡбһ…бһ“бҹ’бһӣбҹ„бҹҮбһўбһҗбҹҒбһҡбһ–бһ·бһҸбһ“бһ·бһ„бһ”бһ¶бһ“бһ‘бһ¶бһҷбҹ”

## 5. бһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһ‘бһ¶бһҷбһӣбһ‘бҹ’бһ’бһ•бһӣбһӣбһҫбһҹбҹҶбһҺбһ»бҹҶбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҸбҹҒбһҹбҹ’бһҸ

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҠбҹӮбһӣбһҷбһҫбһ„бһ”бһ¶бһ“бһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбһҳбҹүбһјбһҢбҹӮбһӣбһҳбһҪбһҷбһҡбһҪбһ…бһ бһҫбһҷ бһҷбһҫбһ„бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһңбһ¶бһҠбһҫбһҳбҹ’бһ”бһёбһ‘бһ¶бһҷбһҖбһ¶бһҡбһңбһ·бһңбһҢбҹ’бһҚбһҮбҹҶбһ„бһә y бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҹбҹҶбһҺбһ»бҹҶбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҸбҹҒбһҹбҹ’бһҸбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ [parsnip::predict()](https://parsnip.tidymodels.org/reference/predict.model_fit.html)бҹ” бһңбһ¶бһ“бһ№бһ„бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бҹ’бһҡбһҫбһҠбһҫбһҳбҹ’бһ”бһёбһӮбһјбһҡбһ”бһ“бҹ’бһ‘бһ¶бһҸбҹӢбһҡбһңбһ¶бһ„бһҖбҹ’бһҡбһ»бһҳбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбҹ”


```r
# Make predictions for the test set
predictions <- lm_mod %>% 
  predict(new_data = diabetes_test)

# Print out some of the predictions
predictions %>% 
  slice(1:5)
```

Woohoo! рҹ’ғрҹ•ә бһҷбһҫбһ„бһ‘бһҫбһ”бһҸбҹӮбһ”бһҺбҹ’бһҸбһ»бҹҮбһҳбҹүбһјбһҠбҹӮбһӣбһҳбһҪбһҷ бһ бһҫбһҷбһ”бҹ’бһҡбһҫбһңбһ¶бһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһ–бҹ’бһҷбһ¶бһҖбһҡбһҺбҹҚ!

бһ“бҹ…бһ–бҹҒбһӣбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһ–бҹ’бһҷбһ¶бһҖбһҡбһҺбҹҚ бһ‘бһҳбҹ’бһҡбһ„бҹӢбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҡбһ”бһҹбҹӢ tidymodels бһӮбһәбһҸбҹӮбһ„бһҸбҹӮбһ”бһ„бҹ’бһҖбһҫбһҸбһҸбһ¶бһҡбһ¶бһ„ tibble/data frame бһ“бҹғбһӣбһ‘бҹ’бһ’бһ•бһӣбһҠбҹӮбһӣбһҳбһ¶бһ“бһҲбҹ’бһҳбҹ„бҹҮбһҮбһҪбһҡбһҲбһҡбһ”бҹ’бһҡбһҖбҹ’бһҡбһҸбһёбҹ” бһ“бҹҒбҹҮбһ’бҹ’бһңбһҫбһІбҹ’бһҷбһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбһҫбһҳ бһ“бһ·бһ„бһҖбһ¶бһҡбһ–бҹ’бһҷбһ¶бһҖбһҡбһҺбҹҚбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ‘бһҳбҹ’бһҡбһ„бҹӢбһҠбҹӮбһӣбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһ”бһ¶бһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһ”бһ“бҹ’бһҸбһҠбһјбһ…бһҮбһ¶бһҖбһ¶бһҡбһӮбһјбһҡбһҡбһјбһ”бҹ”

`dplyr::bind_cols()` бһ—бҹ’бһҮбһ¶бһ”бҹӢбһҮбһҪбһҡбһҲбһҡбһ–бһ бһ»бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ”бһ¶бһ“бһҷбҹүбһ¶бһ„бһҳбһ¶бһ“бһ”бҹ’бһҡбһҹбһ·бһ‘бҹ’бһ’бһ—бһ¶бһ–бҹ”


```r
# Combine the predictions and the original test set
results <- diabetes_test %>% 
  bind_cols(predictions)


results %>% 
  slice(1:5)
```

## 6. бһҡбһјбһ”бһ—бһ¶бһ–бһӣбһ‘бҹ’бһ’бһ•бһӣбһҳбҹүбһјбһҠбҹӮбһӣ

бһҘбһЎбһјбһңбһ“бҹҒбҹҮ бһ–бҹҒбһӣбһңбҹҒбһӣбһ¶бһҠбһҫбһҳбҹ’бһ”бһёбһҳбһҫбһӣбһңбһ¶ secara visual рҹ“Ҳбҹ” бһҷбһҫбһ„бһ“бһ№бһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ– scatter plot бһҡбһ”бһҹбҹӢбһҸбһҳбҹ’бһӣбҹғ `y` бһ“бһ·бһ„ `bmi` бһ‘бһ¶бҹҶбһ„бһўбһҹбҹӢбһҖбҹ’бһ“бһ»бһ„бһҹбҹҶбһҺбһ»бҹҶбһҸбҹҒбһҹбҹ’бһҸ бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһ”бҹ’бһҡбһҫбһҖбһ¶бһҡбһ–бҹ’бһҷбһ¶бһҖбһҡбһҺбҹҚбһҠбһҫбһҳбҹ’бһ”бһёбһӮбһјбһҡбһ”бһ“бҹ’бһ‘бһ¶бһҸбҹӢбһ“бҹ…бһҖбһ“бҹ’бһӣбҹӮбһ„бһҠбҹӮбһӣбһҹбһҳбһҹбҹ’бһҡбһ”бһ”бҹҶбһ•бһ»бһҸ бһ…бһ“бҹ’бһӣбҹ„бҹҮбһҖбҹ’бһҡбһ»бһҳбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ“бҹғбһҳбҹүбһјбһҠбҹӮбһӣбҹ”

R бһҳбһ¶бһ“бһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбҹ’бһҡбһ¶бһ бҹ’бһңбһ·бһ… бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮ `ggplot2` бһӮбһәбһҮбһ¶бһҜбһҖбһҜбһҖбһ—бһ¶бһ–бһҳбһҪбһҷбһҖбҹ’бһ“бһ»бһ„бһ…бҹҶбһҺбҹ„бһҳбһҹбҹ’бһҡбһҹбҹӢбһҹбҹ’бһўбһ¶бһҸ бһ“бһ·бһ„бһ…бһҳбҹ’бһҡбһ»бҹҮбһ”бҹҶбһ•бһ»бһҸбҹ” бһңбһ¶бһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһІбҹ’бһҷбһўбҹ’бһ“бһҖбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбҹ’бһҡбһ¶бһ бҹ’бһңбһ·бһ…бһҠбҹ„бһҷ **бһ”бҹ’бһҡбһҳбһјбһӣбһ•бҹ’бһҸбһ»бҹҶбһ’бһ¶бһҸбһ»бһҠбҹ„бһҷбһҜбһҖбһҡбһ¶бһҮбҹ’бһҷ**бҹ”
